1a.

The ring is pinned at the pivot point on its rim. The table is frictionless, so the only external force acts at the pivot, meaning it produces no torque about the pivot. Therefore angular momentum is conserved:

$$\vec{L}_{total} = \text{constant}$$

Since the system starts at rest, therefore:

$$\vec{L}_{total} = 0 \quad \text{(at all times)}$$

Using the parallel axis theorem, shifting from the ring's center (distance $R$ to pivot), the moment of intertia is:

$$I_p = I_{center} + m_1 R^2 = m_1 R^2 + m_1 R^2 = 2m_1 R^2$$

When the bug is halfway around, it is at distance $2R$ from the pivot. Conservation of angular momentum gives:

$$L_{ring} + L_{bug} = 0$$

$$I_p \omega + m_2 v_{abs} (2R) = 0$$

The bug's absolute velocity is his speed relative to the ring minus the ring surface velocity at $2R$:

$$v_{abs} = v - 2R\omega$$

Substituting and solving for $\omega$:

$$2m_1 R^2 \omega + m_2(v - 2R\omega)(2R) = 0$$

$$\omega = -\frac{m_2 v}{R(m_1 + 2m_2)} \hat{k}$$

The negative sign indicates the ring rotates opposite to the bug (clockwise if bug walks counterclockwise).

1b. 

When the bug returns to the pivot, his distance from the pivot is $r = 0$, so:

$$L_{bug} = m_2 v (0) = 0$$

Since angular momentum is conserved and $L_{total} = 0$:

$$L_{ring} = 0 \implies I_p \omega = 0$$

Since $I_p = 2m_1 R^2 \neq 0$:

$$\boxed{\omega = 0}$$

The ring returns to rest when the bug returns to the pivot.

2a.

This is a perfectly inelastic collision (particles stick to rod), so kinetic energy is not conserved.
However, since the pivot exerts no torque about P, angular momentum is conserved:

$$L_i = L_f$$

Using the parallel axis theorem, with the center of the rod at distance $d/2$ from P:

$$I_{rod,P} = \frac{1}{12}md^2 + m\left(\frac{d}{2}\right)^2 = \frac{1}{3}md^2$$

Adding the two particles as point masses at their respective distances from P:

$$I_{total} = \frac{1}{3}md^2 + m\left(\frac{2d}{3}\right)^2 + m\left(\frac{d}{3}\right)^2 = \frac{8}{9}md^2$$

Particle 1 moves right at $2d/3$ from P (clockwise, positive), particle 2 moves left at $d/3$ from P (counterclockwise, negative):

$$L_i = mv\left(\frac{2d}{3}\right) - mv\left(\frac{d}{3}\right) = \frac{mvd}{3}$$


$$L_i = L_f \implies \frac{mvd}{3} = \frac{8}{9}md^2 \omega_f$$

$$\omega_f = \frac{3v}{8d}$$

2b.

Initial kinetic energy (two particles, both speed $v$):

$$K_i = \frac{1}{2}mv^2 + \frac{1}{2}mv^2 = mv^2$$

Final kinetic energy (everything rotating together at $\omega_f$):

$$K_f = \frac{1}{2}I_{total}\omega_f^2 = \frac{1}{2} \cdot \frac{8}{9}md^2 \cdot \left(\frac{3v}{8d}\right)^2 = \frac{mv^2}{16}$$

Therefore:

$$\frac{K_f - K_i}{K_i} = \frac{\frac{mv^2}{16} - mv^2}{mv^2} = -\frac{15}{16}$$

The system loses 15/16 of its initial kinetic energy to heat and deformation. This large loss is expected for a perfectly inelastic collision; the particles stick to the rod and partially work against each other (one pushing clockwise, one counterclockwise), so most of their kinetic energy is dissipated rather than converted to useful rotational motion.

3.

The collision is elastic so both angular momentum about P and kinetic energy are conserved.
Linear momentum is NOT conserved; the pivot exerts a force at P during the collision.

The two unknowns are $v_f$ (ball rebound speed) and $\omega_f$ (rod angular velocity).

Using parallel axis theoream, moment of inertia of the rod about $P$ is:

$$I_p = \frac{1}{12}md^2 + m\left(\frac{d}{2}\right)^2 = \frac{1}{3}md^2$$

By conservation of angular momentum..

Ball hits at $d/2$ from P. Ball rebounds leftward (negative), rod rotates clockwise (positive):

$$mv_i\frac{d}{2} = -mv_f\frac{d}{2} + \frac{1}{3}md^2\omega_f$$

Simplifying:

$$v_i + v_f = \frac{2d}{3}\omega_f \quad \text{...(1)}$$

By conservation of kinetic energy...

$$\frac{1}{2}mv_i^2 = \frac{1}{2}mv_f^2 + \frac{1}{2}\cdot\frac{1}{3}md^2\omega_f^2$$

Simplifying:

$$v_i^2 - v_f^2 = \frac{d^2}{3}\omega_f^2 \quad \text{...(2)}$$

We can simplify by..

Dividing (2) by (1) using $v_i^2 - v_f^2 = (v_i+v_f)(v_i-v_f)$:

$$v_i - v_f = \frac{d}{2}\omega_f \quad \text{...(3)}$$

Adding (1) and (3):

$$2v_i = \frac{7d}{6}\omega_f$$

$$\omega_f = \frac{12v_i}{7d}$$

4.

Elastic collision; angular momentum about P and kinetic energy are conserved.
Linear momentum is NOT conserved (pivot exerts force at P).
Two unknowns: $v_f$ (object rebound speed) and $\omega_f$ (hoop angular velocity).

Using parallel axis theorem, with center at distance $R$ from P:

$$I_p = m_R R^2 + m_R R^2 = 2m_R R^2$$

Using conservation of momentum, object hits at distance $R$ from P. Object rebounds left (negative), hoop rotates CCW (positive):

$$mv_0 R = -mv_f R + 2m_R R^2 \omega_f$$

Dividing by $R$:

$$mv_0 = -mv_f + 2m_R R\omega_f \quad \text{...(1)}$$

Using conservation of energy:

$$\frac{1}{2}mv_0^2 = \frac{1}{2}mv_f^2 + \frac{1}{2}(2m_R R^2)\omega_f^2$$

$$m(v_0 + v_f)(v_0 - v_f) = 2m_R R^2 \omega_f^2 \quad \text{...(2)}$$

To eliminate $\omega_f$ we divide (2) by (1):

$$v_0 - v_f = R\omega_f \quad \text{...(3)}$$

Substituting (3) into (1):

$$m(v_0 + v_f) = 2m_R(v_0 - v_f)$$

Solving for $v_f$:

$$\boxed{v_f = v_0\left(\frac{2m_R - m}{2m_R + m}\right)}$$

5.

First we find $v_1$ (circular orbit)

Gravity provides centripetal force:

$$\frac{Gm_pm_1}{R^2} = \frac{m_1v_1^2}{R} \implies v_1 = \sqrt{\frac{Gm_p}{R}}$$

Then we find $v_p$ (elliptical orbit)

Gravity is radial so angular momentum is conserved for spaceship 2:

$$m_2v_2(3R) = m_2v_p(R) \implies v_p = 3v_2$$

Conservation of energy between $3R$ and $R$:

$$\frac{1}{2}m_2v_2^2 - \frac{Gm_pm_2}{3R} = \frac{1}{2}m_2v_p^2 - \frac{Gm_pm_2}{R}$$

Substituting $v_p = 3v_2$ and solving:

$$v_p = \sqrt{\frac{3Gm_p}{2R}}$$

Since $v_p > v_1$, spaceship 2 must **slow down** to dock:

$$\Delta v = v_1 - v_p = \sqrt{\frac{Gm_p}{R}}\left(1 - \sqrt{\frac{3}{2}}\right)$$
