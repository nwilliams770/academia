1. 

No slipping between the blocks means that $a_{m1}= a_{m2}$ and $f_{static} \leq \mu_{s}N$.

Because we are looking for the maximum force without the blocks slipping we know that at $F_{max}$, $f_{static} = \mu_{s}N$.

Now we can look at the free-body diagrams for $m1$ and $m2$.

For $m1$:

$F - f_{static} = m_1a$

For $m2$:

$f_{static} = m2_a$

From here we can solve for $a$ and resubstitute:

$F - m2_a = m_1a$ 

$F = m_1a + m2_a = a(m1 + m2)$

$a = \frac{F}{(m1 + m2)}$

$f_{static} = m2 \frac{F}{(m1 + m2)}$ 

This is the friction needed to keep $m2$ moving with $m1$.

Now we must calculate the maximum friction. 

We can derive the normal force, $N$ by starting with $F_{y} = m_{2}a_{y}$:

We know that $a_{y} = 0$ (The block is not accelerating in the vertical direction) and the forces are $N$ (Normal force between $m1$ and $m2$) and $m1g$.

Therefore:
$N - m1g = 0$
$N = m1g$

So the maximum friction is $f_{static} = \mu_{s}m1g$.

We know that the required friction must equal the max friction so:
$m2 \frac{F}{(m1 + m2)} = \mu_{s}m1g$

We solve for $F$ and plug in the provided values:
$F = \mu_{s}m1g(\frac{m1}{m2} + 1) = 15.7 N$

2b.

Knowing the magnitudes of the vector components of the tensile force $T$ allows to calculate it:
$$
T = \sqrt{(Mg)^2 + (Ma)^2} = M \sqrt{g^2 + a^2}
$$

Now that know all the forces we can solve for θ by eliminating the unknown $T$. We can do this by taking the ratio of the parallel force equation to the perpendicular force equation:
$$
\tan \theta = \frac{T \sin \theta}{T \cos \theta} = \frac{Ma - Mg \sin \alpha}{Mg \cos \alpha} = \frac{a - g \sin \alpha}{g \cos \alpha}
$$

$$
\theta = \arctan(\frac{a - g \sin \alpha}{g \cos \alpha}) 
$$
