1a.

We know that the acceleration of the car as a function of time is...
$$a_{car} = \begin{cases} 0 &0\leq t \leq t_1 \\ -c(t-t_1)& t_1<t\leq t_2 \end{cases}.$$

To find the velocity of the car as a function of time, $v_{car}(t)$ we must integrate!
$$v_{car}(t) = v_{car}(0) + \int_{0}^{t} a_{car}dt' \ = \begin{cases} v_0 & 0 \leq t \leq t_1 \\ v_0 - c/2 (t - t_{1})^2 & t_1 \leq t \leq t_2 \end{cases}$$

To find the position of the car as a function of time, $x_{car}(t)$, we integrate again!

For the first phase, $0 \leq t \leq t_1$, we know that velocity is constant so the function is linear (imagine we plot the velocity graph, we're just calculating the area of the rectangle).

But we can also back to the Fundamental Theorem of Calculus (FTC) to logicize this intution:
$$x_{car}(t) - x_{car}(0) =  \int_{0}^{t} v_{car}(t')dt'$$

$x_{car}(0)$ in our system is the origin and $v_{car}(t')$ for this interval is just $v_0$ therefore 
$$x_{car}(t) =  \int_{0}^{t} v_0dt' = v_0(t - 0) = v_0t $$ 

For the second phase, we can also start with the FTC: 
$$x_{car}(t) - x_{car}(t1) =  \int_{t1}^{t} v_{car}(t')dt' $$

therefore
$$x_{car}(t) = x_{car}(t1) + \int_{t1}^{t} v_{car}(t')dt' $$

Luckily, we already solved for phase two's $v_{car}(t')$ previously so we can just plug that in...
$$x_{car}(t) = x_{car}(t1) + \int_{t1}^{t} (v_0 - c/2 (t' - t_{1})^2)dt' $$

Next split up that integral using linearity and pull out the constant
$$x_{car}(t) = x_{car}(t1) + \int_{t1}^{t} v_0dt' - c/2 \int_{t1}^{t} (t' - t_{1})^2dt' $$

Now we can integrate!
$$x_{car}(t) = x_{car}(t1) + v0(t - t1) - c/6(t - t_{1})^3$$

1b.

Knowing the initial velocity of the car, $v0$, $t1$, and the acceleration, $c$, we can use the previous equation to find the value of $t2$ (this is the time where the car is fully at rest, $v_{car}(t2) = 0$

$$ v_{car}(t2) = 0 = 12 - 6/2(t2 - 1)^2 = 12 - 3(t2 - 1)^2 $$
$$ t2 = 3 $$

Next we need to find $\Delta x_{bike}$, we can do this using $\Delta x_{car}$ (we know they meet at $t2=3$) and the fact that the bike was 17m behind the car at $t1$. 

First let's calculate $\Delta x_{car}$ using our previous formula:

$$ x_{car}(t2) = v0t2 - c/6(t2 - t1)^3 = 12 * 3 - 6/6(3 - 1)^3 = 28 $$

Using the initial displacement of the bike, we can calculate $\Delta x_{bike} = \Delta x_{bike}(t2) - \Delta x_bike(t0) = 28 - -17 = 45 $

Then we can calculate $v_{bike}$

$$ v_{bike} = \Delta x / \Delta t = 45 / 3 = 15 \ m/s $$

3.

The rocket starts from rest and its initial position is zero.

Integrate the provided acceleration function to find its velocity function, then integrate again to get its position function.

We can start with the fundamental theorem of calculus again:
$$
v(t) - v(0) = \int_{0}^{t}a(t')dt' = \int_{0}^{t}(A - Bt')dt' = At - B/2t^2 + C
$$

We know that $v(0) = 0$, therefore $C = 0$ and we can write $v(t)$ as $At - B/2t^2$.

We integrate again to find the position function.
$$
y(t) - y(0) = \int_{0}^{t}v(t')dt' = \int_{0}^{t}(At' - B/2t'^2)dt' = A/2t^2 - B/6t^3 + C
$$

We also know that $y(0) = 0$ therefore $C = 0$ and we can write $y(t)$ as $A/2t^2 - B/6t^3$.

Now we can solve for the $t$, $t_{max}$ when the rocket has reached its maximum height e.g. when it made contact with the pebble. We know that at this $t_{max}$, $v = 0$:
$$
v(t_{max}) = 0 = At_{max} - B/2t_{max}^2
$$
$$
t_{max} = 2A/B
$$

Find $y(t_{max})$ to get the max height the rocket reached:
$$
y(2A/B) = 2A^3 / 3B^2
$$

For the $y(t)$ of the pebble, we know that it's falling with an acceleration of $-g$ starting from height $h$. Integrating we get:
$$
y(t) = h - g/2t^2
$$

The pebble and rocket meet when the rocket reaches maximum height therefore:
$$
h - g/2(2A/B)^2 = 2A^3 / 3B^2
$$

Solve for $h$
$$
h = \frac{A^2}{3B^2}(2A + 6g)
$$

First let's get the vertical position function for the ball. We know that $a_{y} = -g$, $v_{y}(0) = v_{0}sin(\theta)$, $y(0) = 0$.

A refresher...we use basic trig to split $v_{0}$ into its horizontal and vertical components: $sin(\theta) = \frac{opposite}{hypoteneuse} = \frac{v_{0y}}{v_{0}}$

We use the fundamental theorem of calculus (FTC) to derive $y(t)$, noting the starting velocity:
$$
y(t) = v_{0}sin(\theta)t - \frac{gt^2}{2} 
$$

Now we need to find when the call was caught $T$ which is when $y(T) = 0$ and $T = 0$:
$$
T = \frac{2v_{0}sin(\theta)}{g}
$$

The horizontal component of the ball's position is $x(t) = v_{0}cos(\theta)t$.

We can use FTC to get the position function of the runner:
$$
x(t) = d + \frac{Bt^3}{6}
$$

At this catch time $T$, $x_{ball}(T) = x_{runner}(T)$, we can plug in $T$ then solve for $B$:
$$
B = g(\frac{cos \theta}{sin \theta}) - \frac{dg^2}{2v_{0}^2sin^2\theta}
$$

5.

Firstly, we assume that the second can is thrown exactly 4.0 seconds after the first and that the initial velocity $v_0$ of each can is the same.

The position function of the first can is:
$$
y_1(t) = v_0t - \frac{gt^2}{2}
$$

The position function of the second can is:
$$
y_2(t) = v_0(t - 4) - \frac{g(t - 4)^2}{2}
$$

At time $t_{collision}$, $y_1(t_{collision}) = y_2(t_{collision})$, we can expand and solve for $t_{collision}$ in terms of $v_0$:
$$
t_{collision} = \frac{v_0}{g} + 2
$$

We know that
$$
y_1(t_{collision}) = 5 = v_0t_{collision} - \frac{gt_{collision}^2}{2}
$$

So we can just plug in $t_{collision}$, solve for $v_0$, then back-substitute and solve for $t_collision$
$$
v_0 = \sqrt{4g^2 + 10g} = 21.98 
$$
