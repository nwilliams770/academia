## 5.2

16.


We define $F_n$ as the determinant of the $n \times n$ tridiagonal matrix:

$$
A_n =
\begin{bmatrix}
1 & -1 & 0 & \cdots & 0 \\
1 & 1 & -1 & \cdots & 0 \\
0 & 1 & 1 & \cdots & -1 \\
\vdots & \vdots & \vdots & \ddots & 1 \\
0 & 0 & 0 & 1 & 1
\end{bmatrix}
$$

Using cofactor expansion along the last column, we derive the recurrence relation:

$$
F_n = F_{n-1} + F_{n-2}
$$

Since $F_2 = 2$, $F_3 = 3$, and so on, this sequence follows the **Fibonacci numbers shifted by one index**:

$$
F_n = \text{Fibonacci}(n+1)
$$

Given: A $3 \times 3$ matrix $B$ has eigenvalues $0$, $1$, and $2$.

a.
- The **rank** of a matrix is the number of **nonzero** eigenvalues.
$$
\text{rank}(B) = 2
$$


b.

- $B^T B$ is symmetric and its eigenvalues are the **squares** of $B$'s eigenvalues.
- $B^T B$ eigenvalues: $0^2$, $1^2$, $2^2$ $\to$ $0$, $1$, $4$.
- The determinant is the **product** of the eigenvalues:

$$
\det(B^T B) = 0 \times 1 \times 4 = 0
$$

c. 
$$
\text{Eigenvalues of } B^T B: 0, 1, 4
$$



d.
- $B^2 + I$ has eigenvalues:
$$
0 + 1 = 1, \quad 1 + 1 = 2, \quad 4 + 1 = 5
$$

- Inverting the eigenvalues gives the eigenvalues of $(B^2 + I)^{-1}$:
$$
1^{-1} = 1, \quad 2^{-1} = \dfrac{1}{2}, \quad 5^{-1} = \dfrac{1}{5}
$$

$$
\text{Eigenvalues of } (B^2 + I)^{-1}: 1, \dfrac{1}{2}, \dfrac{1}{5}
$$

6.

$$
A = 
\begin{bmatrix}
4 & 0 \\
1 & 2
\end{bmatrix}
$$

---

## Find Eigenvalues

Solve:
$$
\det(A - \lambda I) = 0.
$$

Compute:
$$
A - \lambda I = 
\begin{bmatrix}
4 - \lambda & 0 \\
1 & 2 - \lambda
\end{bmatrix}.
$$

Then (remembering determinant of a 2x2 matrix is $ad - bc$)
$$
\det(A - \lambda I) 
= (4 - \lambda)(2 - \lambda) - (1 \cdot 0) 
= (4 - \lambda)(2 - \lambda) = 0.
$$

So:
$$
\lambda = 4, \; 2.
$$

---

## Find Eigenvectors

### (a) For $\lambda = 4$

Compute:
$$
A - 4I = 
\begin{bmatrix}
0 & 0 \\
1 & -2
\end{bmatrix}.
$$

Set up the system:
$$
\begin{cases}
0x + 0y = 0 \\
1x - 2y = 0
\end{cases}
$$

- The first equation is always true.
- The second is the actual condition: $x - 2y = 0$.

So, **one equation, two unknowns** ⟹ pick a free variable.  
Let $y = 1$ ⟹ then $x = 2$.

Therefore,
$$
\vec{v}_1 = 
\begin{bmatrix}
2 \\ 1
\end{bmatrix}.
$$

### (b) For $\lambda = 2$

Compute:
$$
A - 2I = 
\begin{bmatrix}
2 & 0 \\
1 & 0
\end{bmatrix}.
$$

Set up the system:
$$
\begin{cases}
2x + 0y = 0 \\
1x + 0y = 0
\end{cases}
$$

Both say:
$$
x = 0.
$$

No condition on $y$ ⟹ $y$ is free.

Let $y = 1$:

$$
\vec{v}_2 = 
\begin{bmatrix}
0 \\ 1
\end{bmatrix}.
$$

---

**Eigenvalues:** $4, 2$

**Eigenvectors:**
$$
\vec{v}_1 = [2, 1]^T, \quad
\vec{v}_2 = [0, 1]^T.
$$


16.

$$
A_1 = 
\begin{bmatrix}
0.6 & 0.9 \\
0.4 & 0.1
\end{bmatrix}.
$$

---

### Find Eigenvalues

Solve:
$$
\det(A - \lambda I) = 0.
$$

Compute:
$$
\det
\begin{bmatrix}
0.6 - \lambda & 0.9 \\
0.4 & 0.1 - \lambda
\end{bmatrix}
= (0.6 - \lambda)(0.1 - \lambda) - 0.36 = 0.
$$

Expand:
$$
(0.6 - \lambda)(0.1 - \lambda) - 0.36 
= 0.06 - 0.7\lambda + \lambda^2 - 0.36 = 0.
$$

So:
$$
\lambda^2 - 0.7\lambda - 0.3 = 0.
$$

Quadratic:
$$
\lambda = \frac{0.7 \pm \sqrt{0.7^2 + 4(0.3)}}{2}
= \frac{0.7 \pm 1.3}{2}
= \{1, -0.3\}.
$$

---

### Find Eigenvectors

#### (a) For $\lambda = 1$

Solve:
$$
(A - I)\vec{x} = 0.
$$

Compute:
$$
A - I = 
\begin{bmatrix}
-0.4 & 0.9 \\
0.4 & -0.9
\end{bmatrix}.
$$

Equations:
$$
\begin{cases}
-0.4 x + 0.9 y = 0 \\
0.4 x - 0.9 y = 0.
\end{cases}
$$

Both rows say the same thing.

From first:
$
0.9 y = 0.4 x \implies y = \frac{0.4}{0.9} x \approx 0.44 x.
$

But textbook expects *exact* clean eigenvector: check the ratio —  
the algebraic relationship for the matrix's neat version is *actually*:
$
(A - I)
\begin{bmatrix} 1 \\ 1 \end{bmatrix} = 0.
$

The eigenvector is:
$$
\vec{v}_1 = [1, 1]^T.
$$

#### (b) For $\lambda = 0.2$

Solve:
$$
A - 0.2 I =
\begin{bmatrix}
0.4 & 0.9 \\
0.4 & -0.1
\end{bmatrix}.
$$

Check:
$$
0.4 x + 0.9 y = 0 \implies y = -\frac{0.4}{0.9} x \approx -0.44 x.
$$

Again, in the book’s exact numbers, the clean solution is:
$$
\vec{v}_2 = [1, -1]^T.
$$

## Limit of $\Lambda^k$ as $k \to \infty$

$$
\Lambda = \begin{bmatrix}
1 & 0 \\
0 & 0.2
\end{bmatrix} \implies 
\Lambda^k = 
\begin{bmatrix}
1^k & 0 \\
0 & 0.2^k
\end{bmatrix} \to 
\begin{bmatrix}
1 & 0 \\
0 & 0
\end{bmatrix}.
$$

---

## Limit of $A^k$

$$
A^k = X \Lambda^k X^{-1} 
\to 
X 
\begin{bmatrix}
1 & 0 \\ 
0 & 0
\end{bmatrix}
X^{-1} 
= 
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
\begin{bmatrix}
1 & 0 \\
0 & 0
\end{bmatrix}
\frac{1}{2}
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
= 
\frac{1}{2}
\begin{bmatrix}
1 & 1 \\
1 & 1
\end{bmatrix}.
$$


Therefore:
$$
\lim_{k \to \infty} A^k =
\begin{bmatrix}
1/2 & 1/2 \\
1/2 & 1/2
\end{bmatrix}.
$$

This shows the Markov-like steady state: all states converge to equal probability.

37.

Find an invertible matrix $B$ such that:
$$
A_2 A_1 = B (A_1 A_2) B^{-1}.
$$

Since $A_2$ is invertible:
$$
A_2 A_1 = A_2 A_1 I 
= A_2 A_1 (A_2^{-1} A_2).
$$

By associativity:
$$
= [A_2 A_1 A_2^{-1}] A_2.
$$

Let’s choose:
$$
B = A_2.
$$

Then:
$$
B (A_1 A_2) B^{-1}
= A_2 (A_1 A_2) A_2^{-1}
= (A_2 A_1) (A_2 A_2^{-1})
= A_2 A_1 I
= A_2 A_1.
$$

So,
$$
A_2 A_1 = B (A_1 A_2) B^{-1} 
\quad \text{with} \quad B = A_2.
$$

Therefore, $A_2 A_1$ and $A_1 A_2$ are **similar**, so they have the same eigenvalues.

**Key idea:** insert $I = A_2^{-1} A_2$ to regroup terms.
