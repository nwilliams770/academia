1.

The wheel rolls without slipping, giving the constraint:

$$V = R\omega \implies \omega = \frac{V}{R}$$

This means the angle $\theta$ swept by the wheel increases as:

$$\theta(t) = \omega t = \frac{V}{R}t$$

The center of the wheel translates horizontally at constant speed $V$ and stays at fixed height $R$:

$$X(t) = x_0 + Vt, \quad Y(t) = R$$

The bead sits on the rim at distance $R$ from the center, rotating clockwise at rate $\omega$. At $t=0$ the bead is at the top ($\theta = 0$), so:

$$x_{rel} = R\sin\theta, \quad y_{rel} = R\cos\theta$$

When we add the center position and relative position:

$$x(t) = x_0 + Vt + R\sin\left(\frac{V}{R}t\right)$$

$$y(t) = R + R\cos\left(\frac{V}{R}t\right)$$

This curve is called a **cycloid**: the characteristic path traced by a point on the rim of a rolling wheel.

2.

There are two independent sources of motion for the ball's center:

1. The ball being **carried around** by the rotating ring at $\Omega$
2. The ball **rolling** along the ring's surface at angular velocity $\omega$

The center of the ball traces a circle of radius $R - b$ (ring radius minus ball radius).

Because there is no slipping, at the contact point the velocity of the ball's surface relative to the ring is:

$$V' = b(\omega - \Omega)$$

Where $\omega - \Omega$ is the ball's angular velocity **relative to the ring**.

Adding the two contributions:

$$V_{cm} = (R - b)\Omega + V'$$

$$V_{cm} = (R-b)\Omega + b(\omega - \Omega)$$

$$\boxed{V_{cm} = R\Omega + b(\omega - \Omega)}$$

The first term $(R\Omega)$ is the contribution from the ring rotating, the second term $b(\omega - \Omega)$ is the contribution from the ball rolling along the ring's surface.

3a.

Distance from center of mass to ridge corner (by Pythagoras):

$$d = \sqrt{\left(\frac{s}{2}\right)^2 + \left(\frac{s}{2}\right)^2} = \frac{s}{\sqrt{2}}$$

Using parallel axis theorem:

$$I_a = \frac{ms^2}{6} + m\left(\frac{s}{\sqrt{2}}\right)^2 = \frac{ms^2}{6} + \frac{ms^2}{2} = \frac{2ms^2}{3}$$

Before collision, block slides linearly at height $s/2$ above ridge:

$$L_i = mv\frac{s}{2}$$

After collision, block rotates about ridge:

$$L_f = I_a\omega_a = \frac{2ms^2}{3}\omega_a$$

Setting equal and solving:

$$\omega_a = \frac{3v}{4s}$$

3b.

The block just barely doesn't topple when $\omega = 0$ as the center of mass passes directly above the ridge. Using energy conservation post-collision:

$$\frac{1}{2}I_a\omega_a^2 = mg(h_f - h_i)$$

Where:
- $h_i = \frac{s}{2}$ (initial CM height)
- $h_f = \frac{s}{\sqrt{2}}$ (CM height when directly above ridge)

Substituting $I_a = \frac{2ms^2}{3}$ and $\omega_a = \frac{3v_0}{4s}$:

$$\frac{3v_0^2}{16} = \frac{gs}{\sqrt{2}}\left(1 - \frac{\sqrt{2}}{2}\right)$$

$$v_0 = \sqrt{\frac{8gs}{3}\left(\sqrt{2}-1\right)}$$

Any initial speed above this value will cause the block to topple over the ridge.

4.

Three equations govern the motion: translation, rotation, and the no-slip constraint.

**Forces along the incline:**
- $F$ pulls up the incline at the axle
- $f$ is friction at the contact point
- $mg\sin\phi$ is the component of gravity down the incline

**Translation (Newton's second law along incline):**

$$ma = F - f - mg\sin\phi \quad \text{...(1)}$$

**Rotation about center of mass:**

$F$ acts at axle radius $R/3$ (tends to rotate CCW), friction acts at contact radius $R$ (tends to rotate CW):

$$I_0\alpha = fR - \frac{FR}{3} \quad \text{...(2)}$$

**No-slip constraint:**

$$a = R\alpha \quad \text{...(3)}$$

Substituting (3) into (2) with $I_0 = \frac{1}{2}mR^2$:

$$\frac{ma}{2} = f - \frac{F}{3} \quad \text{...(2')}$$

Substituting (1) into (2') and solving:

$$3f + mg\sin\phi = \frac{5F}{3}$$

The yo-yo is on the verge of slipping when $f = \mu_s mg\cos\phi$:

$$3\mu_s mg\cos\phi + mg\sin\phi = \frac{5F_{max}}{3}$$

$$F_{max} = \frac{3mg}{5}(3\mu_s\cos\phi + \sin\phi)$$

5.

The cue delivers a sharp impulse at height $h$ above the centerline. Since the impulse
force is much larger than friction, we ignore friction during the hit. The ball leaves
the cue with forward speed $v_0$ and **backspin** $\omega_0$ (we know it's backspin because
the final speed $\frac{9}{7}v_0 > v_0$, meaning friction acted forward to speed it up —
only possible if the contact point was initially moving backward).


The cue delivers impulse $J = mv_0$ (linear) and $J \cdot h$ (angular about CM):

$$mv_0 h = I_{cm}\omega_0 = \frac{2}{5}mR^2\omega_0$$

$$\omega_0 = \frac{5v_0 h}{2R^2} \quad \text{...(1)}$$


During skidding, friction acts at the contact point. Since friction has zero moment arm
about the contact point, **angular momentum about the contact point is conserved**.

**Initial** (just after cue hit, backspin is negative):

$$L_i = mv_0R - I_{cm}\omega_0$$

**Final** (rolling without slipping at $\frac{9}{7}v_0$, using $\omega_f = \frac{9v_0}{7R}$):

$$L_f = m\frac{9}{7}v_0R + I_{cm}\omega_f$$

Setting $L_i = L_f$:

$$mv_0R - \frac{2}{5}mR^2\omega_0 = m\frac{9}{7}v_0R + \frac{2}{5}mR^2\cdot\frac{9v_0}{7R}$$

Solving for $\omega_0$:

$$\omega_0 = \frac{5v_0}{2R} \quad \text{...(2)}$$


Substituting (2) into (1):

$$\frac{5v_0}{2R} = \frac{5v_0 h}{2R^2}$$

$$\frac{h}{R} = 1$$

The cue must strike the ball exactly at height $R$ above the centerline i.e. at the
very top of the ball, to produce the observed motion.
