2a.

Our system is the single astronaut and the ball. No external forces act on the system so momentum is conserved. 
$$
p_{initial} = p_{final}
$$

where:
$$
p_{initial} = (m_1 + m)v_1
$$

and
$$
p_{final} = m_1 v_{1 \, final} - mv_m
$$

The ball is thrown at speed $u$ relative to the astronaut: $u = v_m + v_{1 \, final} \implies v_m = u - v_{1 \, final}$

Substitute and solve for $v_{1 \, final}$:
$$
v_{1 \, final} = v_1 + \frac{mu}{m1 + m}
$$

2b.

From part a the speed of the ball is
$$
v_m = u - v_{1 \, final} = \frac{m_1 u}{m_1 + m} - v_1
$$

We can apply conservation of momentum to the catch:
$$
m_2 v_2 - mv_m = (m_2 + m) v_{2 \, final} \implies v_{2 \, final} = \frac{m_2 v_2}{m_2 +m} - \frac{m}{m_2 + m} v_m
$$

We want to solve for $v_{1\, final} = v_{2 \, final}$:
$$
\frac{mu}{m_1 + m} + v_1 = \frac{m_2 v_2}{m_2 + m} - \frac{m}{m_2 + m}(\frac{m_1 u}{m_1 + m} - v_1) \implies u = \frac{m_2 (m_1 + m)(v_2 - v_1)}{m(m_2 + m_1 + m)}
$$

4a. 

Start with Newton's second law: $F = \frac{dP}{dt}$

At time $t$, the raindrop has mass $m_r$ and velocity $v_r$ so:
$$
P(t) = m_r v_r
$$

At time $t + \Delta t$, the raindrop has gained mass $\Delta m_r$ and velocity $\Delta v_r$ so:
$$
P(t + \Delta t) = (m_r + \Delta m_r) (v_r + \Delta v_r)
$$

So the change in momentum is:
$$
\Delta P = (m_r + \Delta m_r) (v_r + \Delta v_r) - m_r v_r = m_r \Delta v_r + v_r \Delta m_r + \Delta m_r \Delta v_r
$$

Divide the whole expression by $\Delta t$ and take the limit as $\Delta t \to 0$. Note that the entire $\Delta m_r \Delta v_r$ term shrinks to 0 and vanishes in the limit, leaving:
$$
\frac{dP}{dt} = m_r \frac{dv_r}{dt} + v_r \frac{dm_r}{dt}
$$

At this instant in time, the external force is just gravity which is $g m_r$ so we can subtitute it in. We can also sub in the mass change over time givne by the problem
$$
m_r g = m_r \frac{dv_r}{dt} + v_r (k m_r v_r)
$$

Divide by $m_r$ and isolate to get:
$$
\frac{dv_r}{dt} = g - k v_r^2
$$



4b.

Terminal velocity occurs when acceleration is 0, the drop stops speeding up. Therefore $\frac{dv_r}{dt} = 0$:
$$
0 = g - kv_{terminal}^2 \implies v_{terminal} = \sqrt{\frac{g}{k}}
$$

5a.

The rain falls vertically so it carries no horizontal momentum. There are no external horizontal forces on the vehicle so horizontal momentum is conserved.

They key insight is the mass stays constant at m_0; the rain sticks to the vehicle briefly then leaves so no mass accumulates!

At $t$ the vehicle of mass $m_0$ moves at speed $v$:
$$
P_i = m_0v
$$

After $\Delta t$, rain $\Delta m = b \Delta t$ has come and left, the vehicle now moves at $v + \Delta v$:
$$
P_f = (m_0 + \Delta m)(v + \Delta v)
$$

Since momentum is conserved, $\Delta P = 0$:
$$
0 = (m_0 + \Delta m)(v + \Delta v) - m_0 v = v \Delta m + m_0 \Delta v + \Delta m \Delta v
$$

$\Delta m \Delta v$ vanishes in the limit, and we can sub for $\Delta m$ to solve for $\Delta v$:
$$
0 = vb \Delta t + m_0 \Delta v \implies \Delta v = - \frac{bv}{m_0} \Delta t
$$

The negative sign indicates the vehicle slows down! The rain adds momentum-free mass that the vehicle must drag along.
