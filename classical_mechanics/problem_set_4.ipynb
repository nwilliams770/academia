For block 2 we only care about the j-direction:
$$
2T - m_{2 min} = 0

$$
m_{2 min} = 2T = 2m_1(\sin \theta + \mu_s \cos \theta)

We can find the constraint between the distance block 1 moves up the incline and the distance block 2 moves down by breaking down the total string length into segments:

- Segment A (block 1 to pulley): $L_A = C_1 - x_1$
- Segment B (pulley to pulley): $L_B = C_2 + y_2$
- Segment C (pulley to fixed support) $L_C = C_3 + y_2$

Where $x_1$ is the position of block 1 and $y_2$ is the position of block 2. 

Total length of string: $L_{total} = C_1 - x_1 + C_2 + y_2 + C_3 + y_2$.

$$

a_{x1} = 2g (\frac{m_2 - 2 m_1 \sin \theta - 2 \mu_k m_1 \cos \theta}{4m_1 + m_2})

\frac{dv}{dt} = - \frac{b}{m} e^{cv}
$$

We can solve the differential equation using separation of variables:
$$
int_{v_{0}}^{v}\frac{dv}{e^{cv}}=\int_{0}^{t}-\frac{b}{m}dt

$$
-\frac{1}{c}e^{-cv}+\frac{1}{c}e^{-cv_{0}}=-\frac{b}{m}t
$$

$$
e^{-cv}=\frac{cbt}{m}+e^{-cv_{0}}
$$

$$
|v|=\frac{1}{c}\ln\left(\frac{cbt}{m}+e^{-cv_{0}}\right)

Use differential analysis to take a small slice of the rope: $\Delta r$. This element is distance $r$ from the shaft.

The total mass of the rope is $m1$ and its total length is $l$. Linear mass density (mass per unit length) is $\lambda = \frac{m1}{l}$. Therefore our element $\Delta r$ has mass $\delta m = \lambda \Delta r = \frac{m1}{l} \Delta r$ 

The forces acting on $\Delta r$ are:

- $T(r)$ pulling inward (towards the shaft)
- $T(r + \Delta r)$ pulling outward (away from the shaft)

The element is in circular motion at radius $r$ with angular velocity $\omega$. The centripetral acceleration $a_{centripetal} = \omega^2 r$

Applying Newton's second law:
$$
F_{radial} = T(r) - T(r + \Delta r) = \Delta m \omega^2 r
$$

Substituting $\Delta m$:
$$
T(r) - T(r + \Delta r) = \frac{m1}{l} \Delta r \omega^2 r
$$

Isolate $\Delta r$ to one side then take $\lim_{\Delta r \to 0}$:
$$
\frac{T(r) - T(r + \Delta r)}{\Delta r} = \frac{m1}{l} \omega^2 r
$$

The right side of the equation is unchanged as there is no $\Delta r$, the left side however, is the definition of a derivative (note the sign is flipped though which is logical, the tensile force decreases as we move away from the shaft).

We now have:
$$
-\frac{dT}{dr} = \frac{m1}{l} \omega^2 r
$$

The derivative gives us the rate of change, but we integrate to get the function $T(r)$: the tension value at any position $r$ along the rope.

We can then integrate and simplify:
$$
T(r) = \omega^2 (m_2l + \frac{m1}{2l}(l^2 - r^2))
$$

5a.

The only x-component force is the drag force itself. Apply Newton's second law
$$
F = -C v_x = m a_x
$$

Note the physical implication: as velocity of the object decreases so does the drag force.

Isolate $a_x$:
$$
a_x = -\frac{C v_x}{m}
$$

5b.

The y-direction has two forces: gravitational (downwards, +y) and drag force (resistive upwards, -y).

Apply Newton's second law:
$$
mg - Cv_y = ma_y
$$

Isolate $a_y$:
$$
a_y = g - \frac{Cv_y}{m}
$$

5c/5d.

We can find the $v_x$ and $v_y$ functions of time through integration.

Sub $a_x = \frac{dv_x}{v_x}$, then integrate:
$$
\int_{u}^{v_x} \frac{dv_x}{v_x} = \int_{0}^{t} - \frac{C}{m} dt
$$

$$
\ln(\frac{v_x}{u}) = - \frac{Ct}{m}
$$

Exponentiate and isolate $v_x$:
$$
v_x = u e^{- \frac{Ct}{m}}
$$

Sub $a_y = \frac{dv_y}{v_y}$, then integrate:
$$
\int_{u}^{v_y} \frac{dv_y}{g - \frac{C}{m} v_y} = \int_{0}^{t} dt
$$

$$
-\frac{m}{C} \ln{\frac{g - \frac{C}{m} v_y}{g}} = t
$$

Exponentiate and isolate $v_y$:
$$
v_y = \frac{m}{C}g(1 - e^{-\frac{C}{m}t})
$$

5e/5f.

We can use the velocity functions to determine terminal velocity because terminal velocity is just $\lim_{t\to\infty}$

$v_{xT} = 0$ The horizontal velocity dies off because gravity isn't driving it, only drag is acting, so it keeps slowing until it stops.

$v_{yT} = \frac{mg}{C}$ The vertical velocity reaches a point where gravity and drag perfectly balance.
