From lectures and readings, we've removed the quadratic from conservation of energy and momentum by using difference of squares and relative velocity (relative approach speed before = relative separation speed after):
$$
v_{mf} = \frac{m-M}{m+M}v + \frac{2M}{m+M}u
$$

and

$$
v_{Mf} = \frac{M-m}{M+m}u + \frac{2m}{M + m}v
$$

where $m$ and $M$ have initial velocities $v$ and $u$, respectively.

1.

Because the mass of the pingpong ball is negligible compared to the bowling ball, we drop $m_p$, wherever it's added or subtracted to $m_b$, therefore:
$$
\frac{m_p - m_b}{m_p + m_b} \approx \frac{-m_b}{m_b} = -1
$$

$$
\frac{2m_b}{m_p + m_b} \approx \frac{2m_b}{m_b} = 2
$$

Substituting:
$$
v_{pf} \approx -v_{pi} - 2v_{bi}
$$

2.

Our elastic collision applies here: gravity is a conservative force and the satellite starts and ends far from Earth, so net gravitational PE is unchanged. KE and momentum are conserved. Because the satellite's mass is negligible, we can neglect any effects on the Earth's speed due to gravitation and assume the satellite's momentum is insignificant. 

We can treat the interaction as a one-dimensional elastic collision:
$$
v_{s,f} = \frac{m_s - m_e}{m_s + m_e}4v_e + \frac{2m_e}{m_s + m_e}v_e \approx -2v_e
$$

3a.

At max height, the cube has zero velocity RELATIVE to the block: the move horizontally goether at $v_{b,f}$:
$$
m_c v_{b,i} = (m_c + m_b) v_{b,f} \implies v_{b,f} = \frac{m_c v_{c,i}}{m_c + m_b}
$$

Energy is conserved, but note that some KE is converted to gravitaitonal PE:
$$
\frac{m_c v_{c,i}^2}{2} = \frac{(m_c + m_b) v_{b,f}^2}{2} + m_cgh
$$

Substitute for $v_{b,f}$ and we can solve for $h$:
$$
h = \frac{v_{c,i}^2}{2g} (\frac{m_b}{m_c + m_b})
$$



3b.

Over the entire interaction (cube approaches, slides up, slides down, separates) there is no friction, no external horizontal forces (momentum conserved), cube starts and ends at table height (no net gravitational PE change thus KE is conserved).

This means the whole process is equivalent to an elastic collision between the cube and block, we can apply the general formula directory, noting the initial velocity of the block is 0:
$$
v_{c,f} = \frac{m_b - m_c}{m_c + m_b}v_{c,i}
$$

$$
v_{b,f} = \frac{2m_c}{m_c + m_b}v_{c,i}
$$

4.

Maximum compression occurs when both blocks have zero velocity relative to each other, the momentarily move together with the same velocity.

We know that both energy and momentum are conserved, we can use them to solve for max compression.
$$
m_1v_1 = (m_1 + m_2)v_{12} \implies v_{12} = \frac{m_1 v_1}{m_1 + m_2}
$$

$$
\frac{1}{2}m_1 v_1^2 = \frac{1}{2}(m_1 + m_2)v_{12}^2 + \frac{1}{2}kx^2
$$

Substitute and solve for $x$:
$$
x_{max} = v1 \sqrt{\frac{m_1 m_2}{k(m_1 + m_2)}}
$$

5a.

The collision is perfectly inelastic and the collision instantaneous (spring has no time to act). Only momentum is conserved, some KE is lost.
$$
m_b v_0 = (m_b + M)v_a \implies v_a = \frac{m_b v_0}{M + m_b}
$$

5b.

After the collision, the block and bullet move as a combined mass. The spring force is conservative and the surface is frictionless so energy is conserved!
$$
\frac{1}{2}(M + m_b)v_a^2 = \frac{1}{2}kx_{stop}^2 \implies x_{stop} = \frac{m_b v_0}{\sqrt{k(M + m_b)}}
$$

5c.

$$
\frac{KE_{after}}{KE_{before}} = \frac{ \frac{1}{2}(M + m_b)v_a^2 }{ \frac{1}{2}m_b v_0^2 }
$$

We can substitute the $v_a$ we solved for earlier and simplify to:
$$
\frac{m_b}{M + m_b}
$$

5d.

With friction $\mu_k$ the block travels a total distance $S$, starting and ending at $x=0$. Spring PE therefore cancels. Using conservation of energy:
$$
\frac{1}{2} (M + m_b)v_a^2 = \frac{1}{2}(M + m_b)(\frac{v_a}{2})^2 + \mu_k(M + m_b)gS
$$

We can simplify and solve for $S$:
$$
S = \frac{3v_a^2}{8 \mu_k g}
$$

We are working in 2D so we have conservation momentum equations for both directions, and one for kinetic energy. Note that the masses cancel each other out as we correlate these equations so we omit them here for simplicity.

x-direction:

v_{1,i} = v_{1,f} (\cos \theta_{1,f} + \sin \theta_{1,f})
