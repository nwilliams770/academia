1ai.

$$I = \int r^2 \, dm$$

Every tiny piece of mass $dm$, multiplied by its distance $r$ from the rotation axis, squared. Summed over the whole disc.

We must Turn $dm$ into something we can integrate. The disc has mass $M$ spread uniformly over area $\pi R^2$. So the mass per unit area (surface mass density) is:
$$\sigma = \frac{M}{\pi R^2}$$

Any tiny patch of area $dA$ then has mass $dm = \sigma \, dA$. Substituting:
$$I = \int r^2 \, \sigma \, dA = \sigma \int r^2 \, dA$$

$\sigma$ comes outside because the disc is uniform — density is the same everywhere.

Express $dA$ in polar coordinates. A tiny patch at radius $r$, spanning angle $d\theta$ and radial width $dr$, has sides $dr$ and $r \, d\theta$ (arc length, not just $d\theta$, because angle must be converted to length via arc length $= r\theta$). So:
$$dA = r \, dr \, d\theta$$

Set up the double integral:
$$I = \sigma \int_0^{2\pi} \int_0^R r^3 \, dr \, d\theta$$

We integrate $r$ from $0$ to $R$ and $\theta$ from $0$ to $2\pi$. The double integral just means summing over both variables one at a time — first along a radial slice, then sweeping around the full circle.

Inner integral: 
$$\int_0^R r^3 \, dr = \left[\frac{r^4}{4}\right]_0^R = \frac{R^4}{4}$$

Outer integral:
$\frac{R^4}{4}$ is constant with respect to $\theta$:
$$I = \sigma \cdot \frac{R^4}{4} \cdot \int_0^{2\pi} d\theta = \sigma \cdot \frac{R^4}{4} \cdot 2\pi$$

Substitute $\sigma$:
$$I = \frac{M}{\pi R^2} \cdot \frac{R^4}{4} \cdot 2\pi = \frac{MR^2}{2}$$


$$I_{\text{cm}} = \frac{MR^2}{2}$$

1aii.

We want the moment of inertia about an axis through a point on the rim, a distance $R$ from the center.

Using the Parallel Axis Theorem:

$$I = I_{\text{cm}} + Md^2$$

where $d$ is the distance between the center of mass axis and the new axis. Here $d = R$, so:

$$I_{\text{rim}} = \frac{MR^2}{2} + MR^2 = MR^2\left(\frac{1}{2} + 1\right)$$

$$\boxed{I_{\text{rim}} = \frac{3MR^2}{2}}$$

Moving the axis to the rim means more of the disc's mass is far from the rotation axis, so $I$ increases.

1b.

The washer is symmetric about its central axis, so the center of mass is at the geometric center.

To note:
- The washer has a hole, so $r$ runs from $r_i$ to $r_o$ instead of $0$ to $R$
- It is a 3D object, so $dm = \rho \, dV$ where the volume element in cylindrical coordinates is:

$$dV = r \, dr \, d\theta \, dz$$

So the integral is:

$$I = \rho \int_0^d \int_0^{2\pi} \int_{r_i}^{r_o} r^3 \, dr \, d\theta \, dz$$

Integrating over $z$ just pulls out a factor of $d$ since nothing else depends on $z$. The $\theta$ integral gives $2\pi$. This leaves:

$$I = 2\pi \rho d \int_{r_i}^{r_o} r^3 \, dr = 2\pi \rho d \left[\frac{r^4}{4}\right]_{r_i}^{r_o} = \frac{\pi \rho d}{2}(r_o^4 - r_i^4)$$

Substituting $r_i = 0.0135$ m, $r_o = 0.031$ m, $d = 0.004$ m, $\rho = 7800$ kg m$^{-3}$:

$$I \approx 4.4 \times 10^{-5} \text{ kg m}^2$$

2.

### Step 1 — Newton's 2nd law for the block
$$M_b a = M_b g - T$$

### Step 2 — Torque equation for the pulley
The rope acts on the smaller disc at radius $R/2$:
$$T \cdot \frac{R}{2} = I\alpha$$

### Step 3 — Constraint connecting linear and rotational motion
The rope unwinds from the smaller disc, so linear acceleration of the block equals
the tangential acceleration at radius $R/2$:
$$a = \alpha \cdot \frac{R}{2}$$

### Step 4 — Total moment of inertia
Both discs are solid and rotate together, so we add their moments of inertia:
$$I = \frac{M_1 R^2}{2} + \frac{M_2 (R/2)^2}{2} = \frac{R^2}{8}(4M_1 + M_2)$$

### Step 5 — Solve for acceleration
Substituting $\alpha = 2a/R$ and $I$ into the torque equation:
$$T = \frac{2I\alpha}{R} = \frac{2a}{R^2} \cdot \frac{R^2}{8}(4M_1 + M_2) \cdot 2= 2a\left(M_1 + \frac{M_2}{4}\right)$$

Substituting into the block equation:
$$M_b(g - a) = 2a\left(M_1 + \frac{M_2}{4}\right)$$

$$\boxed{a = \frac{2M_b g}{2M_b + M_2 + 4M_1}}$$

### Step 6 — Time to fall distance $d$
From $d = \frac{1}{2}at^2$:

$$\boxed{t = \sqrt{\frac{d(2M_b + M_2 + 4M_1)}{M_b g}}}$$

**Key pattern:** Newton's 2nd law for linear parts + torque equation for rotating
parts + a constraint linking them. This same recipe applies to every pulley problem.

Take torques about the pivot to eliminate $F_p$. Only the perpendicular component
of each force contributes to rotation:

$$T\sin\beta \cdot L = mg \cdot \frac{L}{2}$$

$L$ cancels:

$$T = \frac{mg}{2\sin\beta} = \frac{4.0 \times 9.8}{2 \times 0.5} = 39.2 \text{ N}$$

Force equilibrium in x and y:

Dividing gives $\tan\alpha = \tan\beta$, so:
$$\alpha = \beta = 30°$$

From either equation above:
$$F_p = T = mg = 39.2 \text{ N}$$

4a.

Force equilibrium perpendicular to slope:
$$N_1 + N_2 = mg\cos\alpha$$

Torque equilibrium about center of mass eliminates friction and gives:
$$N_1 - N_2 = \frac{2mgh\sin\alpha}{d}$$

Solving simultaneously:

$$N_1 = mg\left(\frac{h\sin\alpha}{d} + \frac{\cos\alpha}{2}\right)$$
$$N_2 = mg\left(\frac{\cos\alpha}{2} - \frac{h\sin\alpha}{d}\right)$$

5.

By symmetry both walls are at 45°, so $N_1 = N_2 = N$. Vertical equilibrium:

$$2N\sin 45° = mg$$

$$N = \frac{mg}{\sqrt{2}}$$

Each wall exerts friction $f = \mu N$ at radius $R$, opposing rotation.
Two contact points:

$$\tau_{friction} = 2\mu N R = 2\mu R \cdot \frac{mg}{\sqrt{2}} = \sqrt{2}\mu mgR$$

Constant $\omega_0$ means $\alpha = 0$, so net torque = 0. External torque must
exactly cancel friction:

$$\tau_{ext} = \sqrt{2}\mu mgR$$

The V-groove effectively increases the normal force compared to a
flat surface (where $N = mg$). Here $N = mg/\sqrt{2} > mg/2$ per wall, meaning
more friction and more torque needed to keep it spinning.

6.

Apply Newton's 2nd law for the block (along the slope):
$$ma = mg\sin\beta - T$$

Torque equation for the pulley:
$$TR = I\alpha$$

Constraint linking linear and rotational motion:
$$a = R\alpha$$

Solve for T:
Substituting constraint into torque equation:
$$T = \frac{Ia}{R^2} = \frac{m_p a}{2}$$

Solve for acceleration:
Substituting into block equation:
$$a\left(m + \frac{m_p}{2}\right) = mg\sin\beta$$

$$\boxed{a = \frac{2mg\sin\beta}{2m + m_p}}$$

Time to travel distance $d$:
From $d = \frac{1}{2}at^2$:

$$\boxed{t = \sqrt{\frac{d(2m + m_p)}{mg\sin\beta}}}$$

If $m_p = 0$ we recover pure frictionless incline $a = g\sin\beta$.
The more massive the pulley, the harder it is to spin up, the more it resists
the block, and the longer it takes.
