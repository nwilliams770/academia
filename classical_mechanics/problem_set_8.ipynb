1.

We start by calculating the velocity of the block after decompression using the work-energy theorem:
$$
\frac{mv^2}{2} - \frac{m(0)^2}{2} = \int_{0}^{x} \implies \frac{mv^2}{2} = \frac{kx^2}{2} \implies v = \sqrt{\frac{k}{m}x}
$$

The surface is frictionless, energy is conserved, and we can use this conservation to correlate the time after decompression, $i$, and the time at which the block is on the top of the loop, $f$:
$$
K_{i} = \frac{mv^2}{2}, U_i = 0, K_f = \frac{mv_f^2}{2}, U_f = 2Rmg
$$
$$
\frac{mv^2}{2} = \frac{mv_f^2}{2} + 2Rmg \implies mv_f^2 = mv^2 - 4Rmg
$$

Note that $v$ here is the velocty of the block after decompression as well.

The net force at the top of the hoop is $N + mg$, the normal force is given $N = 2mg$, the centripetal acceleration is $\frac{mv_f^2}{R}$. We can use Newton's second law:
$$
N + mg = 3mg = \frac{mv_f^2}{R}
$$

We previously derived that $mv_f^2 = mv^2 - 4Rmg$ and $v = \sqrt{\frac{k}{m}x}$, we substitute those in and solve for $x$:
$$
x = \sqrt{\frac{7Rmg}{k}}
$$

2a.

Gravity is the only force doing work, total mechanical energy is conserved. We set our reference point where $PE=0$ at the bottom of the circle:
$$
K_{initial} = \frac{mv_0^2}{2}, U_{initial} = 0, K_{final} = \frac{mv_f^2}{2}, U_{final} = 2mgl
$$

Applying conservation of energy we can solve for $v_f^2$:
$$
v_f^2 = v_0^2 - 4gl
$$

At the top of the circle, the tensile force and gravity point toward the center. Together they are the centripetal force required for the motion, therefore:
$$
T + mg = \frac{mv_f^2}{l}
$$

We can substitute for $v_f^2$ and solve for $T$: 
$$
T = \frac{m}{l}(v_0^2 - 5gl)
$$

2b.

The height from the ground is $2l$ and the only vertical force is gravity, therefore the time to hit the ground is:
$$
\frac{1}{2}gt^2 = 2l \implies t = 2 \sqrt{\frac{l}{g}}
$$

We use the horizontal velocity $v_f$ and $t$ to determine $d$:
$$
d = v_f \times t = 2 \sqrt{\frac{l}{g} (v_0^2 - 4gl)}
$$

3.

We have 3 distinct phases: the mass slides from the top to bottom of the ring, the collision at the bottom of the ring, combined objects slide to point A.

The spring length at the top and bottom of the ring are $R \sqrt{5}$ but during this phase, the potential energy is identical at the top and bottom of the ring. We only need to factor in gravity and the motion of the mass using energy conservation:
$$
3mg(2R) = 0 = 0 + \frac{1}{2}(3m)v_1^2 \implies v_1 = 2 \sqrt{Rg}
$$

The collision of the masses is instantaneous so gravity delivers negligible impulse, momentum is conservation:
$$
3mv_1 = 4mv_2 \implies v_2 = \frac{3 \sqrt{Rg}}{2}
$$

As the combined objects move to point A, the spring length changes from $R \sqrt{5}$ at the bottom to $3R$ at point A so the forces do not cancel! At point A we expect $KE = 0$ since the objects are at rest.
$$
\frac{1}{2}4mv_2^2 + \frac{1}{2}k(R \sqrt{5})^2 = 0 + \frac{1}{2}k(3R)^2 + 4mgR
$$

We can substitute for $v_2$ then solve for $m$:
$$
m = \frac{4kR}{g}
$$

4a.

Escape velocity is the min velocity needed to reach infinity with zero KE remaining. Note that we use infinity here because it is our reference point where $U = 0$ for convenience.
The final total energy is 0 therefore:
$$
KE_{initial} + PE_{initial} = 0
$$

We can substitute KE and PE for gravitational forces and solve for $v_0$:
$$
\frac{1}{2}mv_0^2 - \frac{Gm_em}{R_e} = 0 \implies v_0 = \sqrt{2gR_e}
$$

4b.

Since the projectile was launched at escape velocity, its total mechanical energy is zero throughout the journey:
$$
\frac{1}{2}mv_p^2 - \frac{Gm_em}{2R_e} = 0 \implies v_p = \frac{Gm_e}{R_e}
$$

4c.

Gravity provides the energy required to maintain the satellite's circular orbit. We can use the formula for gravitational force and circular motion:
$$
\frac{mv_s^2}{2R_e} = \frac{Gm_em}{(2R_e)^2} \implies v_s = \frac{Gm_e}{2R_e}
$$

4d.

The collision is instantaneous, momentum is conserved, however the projectile and satellite velocities are orthogonal, we treat the momentum as a vector:
$$
mv_p + mv_s = 2mv_c \implies v_c = \frac{v_p + v_s}{2}
$$

We use Pythagoras to find the magnitude, then substitute the velocities we calculated prior
$$
v_c = \frac{v_p^2 + v_s^2}{2} \implies v_c = \frac{3Gm_e}{4R_e}
$$

We use conservation of energy where:
$$
KE_i = \frac{1}{2}mv_0^2, PE_i = U(0) = -U_0 e^0 = -U_0, KE_f = 0, PE_f = U(\pm a) = -U_0 e^{-a^2/a^2} = -U_0 e^{-1}

Plug them in and solve for $v_0$:
$$
