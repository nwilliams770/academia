25.

Find the Cholesky factor $C$ such that  
$$S = C^\top C$$  
for  
$$S = \begin{bmatrix} 4 & 8 \\ 8 & 25 \end{bmatrix}$$

Assume an upper-triangular form for $C$:  
$$C = \begin{bmatrix} a & b \\ 0 & c \end{bmatrix}$$  
Then  
$$
C^\top = \begin{bmatrix} a & 0 \\ b & c \end{bmatrix}
\quad \Rightarrow \quad
C^\top C = \begin{bmatrix} a & 0 \\ b & c \end{bmatrix} \begin{bmatrix} a & b \\ 0 & c \end{bmatrix}
= \begin{bmatrix} a^2 & ab \\ ab & b^2 + c^2 \end{bmatrix}
$$

Match with  
$$S = \begin{bmatrix} 4 & 8 \\ 8 & 25 \end{bmatrix}$$  
So:
$$
a^2 = 4 \Rightarrow a = 2 \\
ab = 8 \Rightarrow b = 4 \\
b^2 + c^2 = 25 \Rightarrow c^2 = 9 \Rightarrow c = 3
$$

$$C = \begin{bmatrix} 2 & 4 \\ 0 & 3 \end{bmatrix}$$

27.


$$
\begin{bmatrix} x & y \end{bmatrix}
\begin{bmatrix}
a & b \\
b & c
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= x^\top S x
$$

This expands to:

$$
ax^2 + 2bxy + cy^2
$$

We complete the square in $x$:

$$
ax^2 + 2bxy + cy^2
= a\left(x + \frac{b}{a} y\right)^2 + \left(c - \frac{b^2}{a}\right)y^2
$$

So the completed square form is:

$$
a\left(x + \frac{b}{a} y\right)^2 + \left(c - \frac{b^2}{a}\right)y^2
$$

This matches the $LDL^\top$ structure where:

- First pivot: $d_1 = a$
- Second pivot: $d_2 = c - \frac{b^2}{a}$

29.

Given
$$
F_1(x, y) = \frac{1}{4}x^4 + x^2 y + y^2
$$

recognize that this expression can be rewritten as a perfect square:

$$
F_1(x, y) = \left( \frac{1}{2}x^2 + y \right)^2
$$

Since squares are always non-negative, we conclude:

- $F_1(x, y) \geq 0$
- The minimum value is $0$
- This occurs when $\frac{1}{2}x^2 + y = 0$, i.e. on the curve:
  
  $$
  y = -\frac{1}{2}x^2
  $$

So $F_1$ has a **global minimum** of $0$ along that curve. No critical point solving or Hessian needed.

**Second derivative matrix $S_1$**:

For completeness, we can compute the Hessian:

$$
S_1 =
\begin{bmatrix}
6x^2 & 2x \\
2x & 2
\end{bmatrix}
$$

At any point, $S_1$ is **positive semidefinite**, confirming the function is convex.

This pattern — recognizing a function as a square — is a fast way to determine convexity and minima in multivariable optimization problems.

3.

#### (a) $T(\mathbf{v}) = (v_2, v_1)$  
Swaps components.  
- $T(\mathbf{v} + \mathbf{w}) = (v_2 + w_2,\ v_1 + w_1) = T(\mathbf{v}) + T(\mathbf{w})$  
- $T(c\mathbf{v}) = (cv_2, cv_1) = cT(\mathbf{v})$  

#### (b) $T(\mathbf{v}) = (v_1, v_1)$  
Duplicates first component.  
- $T(\mathbf{v} + \mathbf{w}) = (v_1 + w_1, v_1 + w_1) = T(\mathbf{v}) + T(\mathbf{w})$  
- $T(c\mathbf{v}) = (cv_1, cv_1) = cT(\mathbf{v})$

#### (c) $T(\mathbf{v}) = (0, v_1)$  
Zeros the first component.  
- $T(\mathbf{v} + \mathbf{w}) = (0, v_1 + w_1) = T(\mathbf{v}) + T(\mathbf{w})$  
- $T(c\mathbf{v}) = (0, cv_1) = cT(\mathbf{v})$

#### (d) $T(\mathbf{v}) = (0, 1)$  
Constant output regardless of input.  
- $T(\mathbf{0}) = (0, 1) \neq \mathbf{0}$

#### (e) $T(\mathbf{v}) = v_1 - v_2$  
Outputs a scalar.  
- $T(\mathbf{v} + \mathbf{w}) = (v_1 + w_1) - (v_2 + w_2) = T(\mathbf{v}) + T(\mathbf{w})$  
- $T(c\mathbf{v}) = c(v_1 - v_2) = cT(\mathbf{v})$

#### (f) $T(\mathbf{v}) = v_1 \cdot v_2$  
Product of components breaks linearity.  
- Try $\mathbf{v} = (1, 0)$ and $\mathbf{w} = (0, 1)$:  
  - $T(\mathbf{v}) = 0$, $T(\mathbf{w}) = 0$, but  
  - $T(\mathbf{v} + \mathbf{w}) = T(1, 1) = 1 \neq 0$

5.

- For vectors **not** on the y-axis ($v_1 \ne 0$), $T(\mathbf{v}) = \mathbf{v}$.
- For vectors **on** the y-axis ($v_1 = 0$), $T(\mathbf{v}) = \mathbf{0}$.


#### Scalar Multiplication Holds

Take any scalar $c$ and any vector $\mathbf{v}$.

- If $\mathbf{v} = (v_1, v_2)$ with $v_1 \ne 0$:  
  $$T(c\mathbf{v}) = c\mathbf{v} = cT(\mathbf{v})$$

- If $\mathbf{v} = (0, v_2)$ (i.e., on the y-axis):  
  $$T(c\mathbf{v}) = T((0, cv_2)) = (0, 0) = c(0, 0) = cT(\mathbf{v})$$

So in all cases, $T(c\mathbf{v}) = cT(\mathbf{v})$.


#### Additivity Fails

Pick two vectors:

- $\mathbf{v} = (1, 2)$ ⇒ $T(\mathbf{v}) = (1, 2)$  
- $\mathbf{w} = (-1, 2)$ ⇒ $T(\mathbf{w}) = (-1, 2)$  
- Then: $\mathbf{v} + \mathbf{w} = (0, 4)$ ⇒ $T(\mathbf{v} + \mathbf{w}) = (0, 0)$

But:  
$$T(\mathbf{v}) + T(\mathbf{w}) = (1, 2) + (-1, 2) = (0, 4)$$  
$$T(\mathbf{v} + \mathbf{w}) = (0, 0) \ne (0, 4)$$

Therefore, $T$ fails to satisfy additivity and is **not linear**.

#### (a) $T(v_1, v_2) = (v_2, v_2)$, with $W = \mathbb{R}^2$

- $T(1, 0) = (0, 0)$ and $T(0, 0) = (0, 0)$  
- So the kernel contains non-zero vectors  
- $\ker(T) \ne \{\mathbf{0}\}$ ⇒ **not injective**


#### (b) $T(v_1, v_2) = (v_1, v_2, v_1 + v_2)$, with $W = \mathbb{R}^3$

- The third output coordinate is always $v_1 + v_2$  
- So the image lies in a plane in $\mathbb{R}^3$:  
  $$\text{Im}(T) = \{ (x, y, x + y) \}$$  
- Example: $(0, 0, 1)$ is **not** in the range

#### (c) $T(v_1, v_2) = v_1$, with $W = \mathbb{R}^1$

- $T(1, 0) = 1$, $T(1, 1) = 1$  
- Different inputs map to same output  
- $\ker(T) = \{(0, v_2)\}$ is infinite

11.

#### (a) What are the input and output spaces $V$ and $W$?

- $V = \mathbb{R}^n$: the domain (input vectors have $n$ components)
- $W = \mathbb{R}^m$: the codomain (output vectors have $m$ components)

#### (b) Why is $\text{range}(T) = \text{column space of } A$?

- $T(\mathbf{v}) = A\mathbf{v}$ is a linear combination of the columns of $A$
- Therefore, the outputs of $T$ span exactly the column space of $A$

$$\text{range}(T) = \text{Col}(A)$$

#### (c) Why is $\text{ker}(T) = \text{nullspace of } A$?

- $\text{ker}(T)$ is the set of vectors sent to zero:  
  $$\text{ker}(T) = \{ \mathbf{v} \in \mathbb{R}^n \mid A\mathbf{v} = \mathbf{0} \}$$
- This is exactly the definition of the nullspace of $A$

$$\text{ker}(T) = \text{Null}(A)$$
