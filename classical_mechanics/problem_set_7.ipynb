Friction removes energy from the system/retards the motion therefore it is negative:
$$
W_{friction} = \int_{0}^{l} - \mu_k mg \cos \theta \, ds \, \hat{i} = - \mu_k mg \cos \theta \int_{0}^{l} ds \, \hat{i} = - \mu_k mg \cos \theta (displacement_{final} - displacement_{initial})
$$

We can plug in the provided values then solve:
$$
-\mu_k mg \cos \theta (x_{final} - x_{initial}) = - 0.2 \times 4.0 \times 9.8 \times \cos{30^\circ} \times (3.0 - 0) \approx -20.4 \, \text{J}
$$

W_g = \int_{0}^{l} mg \sin \theta dx

W_g = mgl \sin \theta = 4.0 \times 9.8 \times 3.0 \times \sin{30^\circ} \approx 58.8 \, \text{J}

1c.

Per the work-energy theorem:
$$
W_{net} = \Delta KE = \frac{1}{2}mv_{final}^2 - \frac{1}{2}mv_{initial}^2
$$

The initial velocity of the block is $0$ so we can eliminate that term. We also know that $W_{net} = W_{friction} + W_g$ which we've solved for above.

First solve for $v_{final}$:
$$
W_{net} = \frac{1}{2}mv_{final}^2 \implies v_{final} = 4.38 \, \text{m/s}
$$

We know that $KE_{initial} = 0$ because the block starts from rest, therefore:
$$
KE = KE_{final} = \frac{1}{2}mv_{final}^2 \approx 38.4 \, \text{J}
$$

1e.

As the block moves along the rough horizontal surface, the only horizontal force is kinetic friction and therefore the only work being done is by friction. We can apply the work-energy theorem, noting that $v_{initial}$ is the velocity at the bottom of the incline we calculated and $v_{final} = 0$ since the object comes to rest.

$$
W_{net} = W_{friction} = \frac{1}{2} mv_{final}^2 - \frac{1}{2}mv_{initial}^2 \implies W_{friction} = \frac{1}{2}mv_{initial}^2
$$

We can calculate the friction work over distance $d$:
$$
W_{friction} = \int_{0}^{d} - \mu_k mg \, ds \, \hat{i} = - \mu_k mg (d_{final} - d_{initial}) = - \mu_k mgd
$$

Set these to equal and solve for $d$:
$$
-\mu_k mgd = -\frac{1}{2}mv_{initial}^2 \implies d = \frac{v_{initial}^2}{2 \mu_k g} \approx 3.26 \, \text{m}
$$

2.

A and B collide on a frictionless surface therefore momentum is conserved. A has velocity $v_A$, B is at rest:
$$
m_A v_A + m_B (0) = (m_A + m_B) v_{final}
$$

Where $v_{final}$ is the post-collision speed of the combined blocks:
$$
v_{final} = \frac{m_A v_A}{m_A + m_B}
$$

This is our *initial* velocity as the combined blocks enter the rough surface at $x = 0$.

Once the blocks hit the rough surface, the only force is the friction, and we want to find the velocity of the blocks where $x=d$ (when the blocks reach the spring).

Using the work-energy theorem:
$$
W_{net} = \Delta KE = \frac{1}{2}(m_A + m_B)v_{final}^2 - \frac{1}{2}(m_A + m_B)v_{initial}^2
$$

The only force doing work on the rough surface is friction and we are given that $\mu_k(x) = bx$, so:
$$
W_{net} = W_{friction} = \int_{0}^{d} -\mu_k(x)(m_A + m_B)gdx = \int_{0}^{d} -\mu_k(x)(m_A + m_B)gdx = -(m_A + m_B)gb \int_{0}^{d} x \, dx = - \frac{(m_A + m_B)gbd^2}{2}
$$

Now we can solve for $v_{final}$:
$$
\frac{1}{2}(m_A + m_B)v_{final}^2 - \frac{1}{2}(m_A + m_B)v_{initial}^2 = - \frac{(m_A + m_B)gbd^2}{2} \implies v_{final} = v_{initial}^2 - bgd^2
$$

The spring is on a frictionless surface so the only force acting on the blocks is the spring force, given by Hooke's law where $k$ is the spring constant (stiffness) and $x$ is the compression distance.
$$
F_{spring} = -kx
$$

We can apply the work-energy theorem again, the blocks come to rest at maximum compression so $v_{final} = 0$. We also know that the $v_{initial}$ is the $v_{final}$ from our previous calculation (when the block has finished moving across the rough surface).
$$
\int_{0}^{x_{max}} -kx \, dx = \frac{1}{2}mv_{initial}^2 - \frac{1}{2}mv_{final}^2
$$

We can substitute $v_{final}$ and $m$ ($m_A + m_B$) to solve for $x_{max}$:
$$
x_{max} = \frac{m_A + m_B}{k} [(\frac{m_A v_A}{m_A + m_B})^2 - bgd^2]
$$

3a.

We know that friction always acts opposite to motion so its work is:
$$
W_{friction} = F_{friction} S
$$
where $S$ is the path length traveled.

From point A to B is half the circumference of the circle therefore $S = \frac{2 \pi R}{2} = \frac{\pi}{R}$ and, because are we on an inclined plane, the normal force is $N = \mu_k mg \cos \phi$. The plane is inclined at $\phi$, gravity's component perpendicular to the plane.

Therefore:
$$
W_{friction} = -\mu_k mg \cos \phi \times \pi R = - \pi R \mu_k mg \cos \phi
$$



3b.

At point B the body is at the top of the circle and the forces acting towards the center are the tension $T$ and the component of gravity along the plane (at point B this points towards hte center).

We can use Newton's second law in the radial direction to determine:
$$
T + mg \sin \phi = \frac{mv_B^2}{R}
$$

To solve for $v_B$ we can use the work-energy theorem using our previous result.
$$
W_{friction} + W_{gravity} = \Delta KE
$$

$$
-\pi R \mu_k mg \cos \phi - 2Rmg \sin \phi = \frac{1}{2} mv_B^2 - \frac{1}{2}mv_0^2 \implies v_B^2 = v_0^2 - 2 \pi R \mu_k g \cos \phi - 4Rg \sin \phi
$$

We can now solve for $T$ then substitue $v_B$:
$$
T = \frac{mv_B^2}{R} - mg \sin \phi = \frac{m}{R}(v_0^2 - 2 \pi R \mu_k g \cos \phi - 4Rg \sin \phi) - mg \sin \phi = \frac{mv_0^2}{R} - 2 \pi \mu_k mg \cos \phi - 5mg \sin \phi
$$
