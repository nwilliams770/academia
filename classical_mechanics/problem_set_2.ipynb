1. 

No slipping between the blocks means that $a_{m1}= a_{m2}$ and $f_{static} \leq \mu_{s}N$.

Because we are looking for the maximum force without the blocks slipping we know that at $F_{max}$, $f_{static} = \mu_{s}N$.

Now we can look at the free-body diagrams for $m1$ and $m2$.

For $m1$:

$F - f_{static} = m_1a$

For $m2$:

$f_{static} = m2_a$

From here we can solve for $a$ and resubstitute:

$F - m2_a = m_1a$ 

$F = m_1a + m2_a = a(m1 + m2)$

$a = \frac{F}{(m1 + m2)}$

$f_{static} = m2 \frac{F}{(m1 + m2)}$ 

This is the friction needed to keep $m2$ moving with $m1$.

Now we must calculate the maximum friction. 

We can derive the normal force, $N$ by starting with $F_{y} = m_{2}a_{y}$:

We know that $a_{y} = 0$ (The block is not accelerating in the vertical direction) and the forces are $N$ (Normal force between $m1$ and $m2$) and $m1g$.

Therefore:
$N - m1g = 0$
$N = m1g$

So the maximum friction is $f_{static} = \mu_{s}m1g$.

We know that the required friction must equal the max friction so:
$m2 \frac{F}{(m1 + m2)} = \mu_{s}m1g$

We solve for $F$ and plug in the provided values:
$F = \mu_{s}m1g(\frac{m1}{m2} + 1) = 15.7 N$

The blocks are connected by an inextensible cord, their velocities are constrained to have the same magnitude. Therefore, if block 2 moves at constant speed ($a_2 = 0$), block 1 must also move at constant speed ($a_1 = 0$), meaning the net force on both blocks is zero.

The top block is being pulled by force $F$ and is resisted by the tensile force in the rope $T$ and the friction between the blocks $f_{between}$:
$$
F = T + f_{between}
$$

The bottom block is pulled by the tensile force in the rope $T$ and is resisted by the friction between the block and the floor $f_{floor}$ and the friction between the blocks $f_{between}$:
$$
T = f_{floor} + f_{between}
$$

We can substite for $T$ and therefore:
$$
F = f_{floor} + 2f_{between}
$$

We can subtitute even further using the Law of Kinetic Friction:
$$
f_{between} = \mu m_2 g
$$

$$
f_{floor} = \mu (m_1 + m_2) g
$$

$$
F = \mu m_2 g + 2\mu (m_1 + m_2) g = \mu g (m_1 + 3m_2)
$$

We know the block is at rest and therefore the forces are balanced (e.g. their magnitudes are equal). The downward force is $mg$ while the spring force is informed by Hooke's law $kd$ where $d$ is the displacement of the spring (e.g. stretch distance).
We can now solve for $d$:
$$
mg = kd
$$
$$
d = \frac{mg}{k}
$$

In figure B, the blocks are connected to the spring via ropes. Therefore the force stretching the spring is equal to the tension in the ropes. We can find the tension $T$ by isolating one of the blocks. The downward force is $\frac{m}{2}g$ and, because the blocks is at rest, the upward tensile force balances it therefore $T = \frac{m}{2}g$.

We also know that the restoring force of the spring equals the tension therefore:
$$
F_{spring} = kx = T = \frac{m}{2}g
$$

Solve the displacement in the spring $x$:
$x = \frac{1}{2} \frac{mg}{k}$

We can substitute $d$ here to determine that:
$x = \frac{d}{2}$

The platform starts at height $h$ when the platform sorts it, and moves to $h_f$ when the pillar is removed. We need to find $\Delta y$. 

The forces acting on the platform are:
* Gravity pulls the platform down at $Mg$
* The bottom springs are compressed by $\Delta y$ and push the platform upward. There are 2 identical springs so the total force is $2k_1 \Delta y$
* The top springs are stretched by $\Delta y$. They pull the platform upward. There are 2 identical springs so the total force is $2k_2 \Delta y$

The platform is at rest so we know the magnitudes of the downward and upward forces are equal:
$$
Mg = 2k_1 \Delta y + 2k_2 \Delta y
$$

Solve for $\Delta y$:
$$
\Delta y = \frac{Mg}{2(k_1 + k_2)}
