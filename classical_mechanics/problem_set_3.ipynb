Fixed $\theta$ means:
- No tangential acceleration along the hoop therefore the forces are balanced
- But the bead moves in a horizontal circle of radius $R \sin \theta$, so it has centripetal acceleration $\omega^2 R \sin \theta$

The normal force is perpendicular to the hoop (along the radius toward the hoop's center).

Vertical force balance:
$$
N \cos \theta - mg = 0
$$

$$
N \cos \theta = mg
$$ 

Horizontal force balance:
$$
N \sin \theta = m \omega^2 R \sin \theta 
$$

We can divide the equations:
$$
\tan \theta = \frac{R \sin \theta \omega^2}{g}
$$

Cancel $\sin \theta$:
$$
\frac{1}{\cos \theta} = \frac{R \omega^2}{g}
$$

Solve for $\omega$:
$$
\omega = \sqrt{frac{g}{R \cos \theta}}
$$

Vertical force balance:
$$
N \cos \theta - mg = 0
$$

Horizontal force balance (centripetal acceleration):
$$
N \sin \theta - m(R \sin \theta) \omega^2 = 0
$$

Isolate and divide to eliminate $N$ and $m$:
$$
\tan \theta = \frac{R \sin \theta \omega^2}{g}
$$

Cancel and sign, then solve for $\omega$:
$$
\omega = \sqrt{\frac{g}{R \cos \theta}}
$$

This tells us which angle $\theta$ can remain in equilibrium for a given spin speed. It is the balanace between gravity pulling downard and rotation pushing the bead outward.

1b.

From part A we derived that:
$$
\cos \theta = \frac{g}{R \omega^2}
$$

Equilibrium angle must satisfy this relation, for a given $\omega$, an equilibrium only exists if the right-hand side equals the cosine of some real angle. 

For real $\theta$: $-1 \leq \cos \theta \leq 1$.

Since $\frac{g}{R \omega^2} > 0$, we only need:
$$
\frac{g}{R \omega^2} \leq 1
$$

If the right-hand side is greater than 1, then no real angle can satisfy the equation. That means no nonzero equilibrium exists (the bead is at the bottom of the hoop).

The **critical** value occurs when the inequality becomes equality:
$$
\frac{g}{R \omega_0^2} = 1
$$

Solve and:
$$
\omega_0 = \sqrt{g/R}
$$

This means that:
- If $\omega < \omega_0$, then $\frac{g}{R \omega^2} > 1$ and no real nonzero $\theta$ exists and only the bottom fo the hoop is equilibrium
- If $\omega > \omega_0$, then $\frac{g}{R \omega^2} < 1$ and a nonzero $\theta$ becomes possible and the bead lifts up

For the car to not be sliding, the friction force $f$ must:
- Be static
- $|f| \leq \mu_{s}N$

At the speed limits we're looking for, friction is maxed out such that: $|f| = \mu_{s}N$.

For $v_{min}$, the vertical forces are:
$$
N \cos \beta + f \sin \beta = mg
$$

We know $f = \mu_{s}$ so we can substitute and solve for $N$
$$
N = \frac{mg}{\cos \beta + \mu_{s} \sin \beta}
$$

For $v_{min}$ we know the horizontal (centripetal) forces are:
$$
N \sin \beta - f \cos \beta = \frac{mv_{min}^2}{R}
$$

Substitute $N$ and solve for $v_{min}$:
$$
v_{min} = \sqrt{Rg (\frac{\sin \beta - \mu_{s} \cos \beta}{\cos \beta + \mu_{s} \sin \beta})}
$$

For $v_{max}$ the vertical forces are: 
$$
N \cos \beta - f \sin \beta - mg = 0
$$

For $v_{max}$ the horizontal forces are:
$$
N \sin \beta + \cos \beta = m \frac{v_{max}^2}{R}
$$

At the limit $f = \mu_{s} N$ so we can substitute and solve for $v_{max}$:
$$
v_{max} = \sqrt{Rg \frac{\sin \beta + \mu_{s} \cos \beta}{\cos \beta - \mu_{s} \sin \beta}}
$$

In the vertical direction, we know the vertical component of $T$ balances $mg$:
$$
T \cos \beta = mg
$$

And the horizontal component of $T$ provides the centripetal force required for circular motion:
$$
T \sin \beta = ma_{radial} = m \omega^2 r
$$

Note $r$ is the radius of the horizontal circular path, not $l$ but we can derive $r$:
$$
r = l \sin \beta
$$

From there we can isolate $T$, plug it in as well as $r$ and solve for $\omega$:
$$
(\frac{mg}{\cos \beta}) \sin \beta = m \omega^2 r
$$

We can simplify the left-hand side, cancel $m$ then plug in $r$, and solve for $\omega$:
$$
g \tan \beta = \omega^2 (l \sin \beta)
$$

$$
\omega = \sqrt{\frac{g}{l \cos \beta}}
$$

5a.

For satellites of mass $m$ at distance $R$ from Earth's center:
$$
\frac{mv^2}{R} = \frac{GM_{Earth}m}{R^2}
$$

so
$$
v = \sqrt{\frac{GM_{Earth}}{R}}
$$

The farther satellite has less velocity as speed would decrease as $R$ increases.

5b.

For period $T$:
$$
T = \frac{2 \pi r}{v} = 2 \pi \sqrt{\frac{r^2}{GM}}
$$

As the period increases with $R$, the farther satellite has the greater period.

6a.

The static friction provides both tangential and radial forces.

Tangential:
$$
F_{tangential} = m \alpha d
$$

Radial:
$$
F_{radial} = m \omega^2 d = m \alpha^2 t^2 d
$$

Total: 
$$
f_s(t) = \sqrt{F_{tangential}^2 + F_{radial}^2} = m \alpha d \sqrt{1 + (\alpha t)^2}
$$
