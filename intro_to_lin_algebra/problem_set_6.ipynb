29.

A unique hyperplane containing the points $0, a_1, \dots, a_{n-1}$ in $ \mathbb{R}^n $ exists if and only if the vectors $a_1, \dots, a_{n-1}$ are **linearly independent**.


Given the matrix:

$$
A = \begin{bmatrix}
| & | & & | \\
a_1 & a_2 & \dots & a_{n-1} \\
| & | & & |
\end{bmatrix}
$$

- If $\text{rank}(A) = n-1$, the columns of $A$ are linearly independent, defining a **unique hyperplane**.
- If $\text{rank}(A) < n-1$, the columns are linearly dependent, meaning a nonzero vector $v$ exists in the **left null space** such that:

$$
v^T x = 0
$$

- In this case, the points already lie in an existing hyperplane, and there are **infinitely many possible hyperplanes**.

If the vectors $a_1, \dots, a_{n-1}$ are dependent, then $v$ is orthogonal to all $a_i$, and the equation $v^T x = 0$ defines the hyperplane through these points.

Given:

$$

Take the dot product with $ q_1 $:
$$
q_1 \cdot (c_1 q_1 + c_2 q_2 + c_3 q_3) = q_1 \cdot 0
$$

By linearity of the dot product:
$$
c_1 (q_1 \cdot q_1) + c_2 (q_1 \cdot q_2) + c_3 (q_1 \cdot q_3) = 0
$$

Since the $ q_i $ are **orthonormal**:
- $ q_1 \cdot q_1 = 1 $
- $ q_1 \cdot q_2 = 0 $, $ q_1 \cdot q_3 = 0 $

Thus:

$$
c_1 = 0
$$

Repeating for $ q_2 $ and $ q_3 $ gives $ c_2 = 0 $ and $ c_3 = 0 $, proving **linear independence**.

$$

Multiply both sides by $ Q^T $:
$$
Q^T Q x = Q^T 0
$$

Since the columns of $ Q $ are **orthonormal**, we have:
$$
Q^T Q = I
$$

Thus:
$$
I x = 0
$$

which simplifies to:

$$
x = 0
$$

$ Q $ has only the trivial null space, meaning its columns are **linearly independent**.

10.

If every row of $A$ sums to zero, the vector of ones:

$$
x = (1,1,1,\dots,1)^T
$$

is in the null space of $A$, meaning $A$ is **singular** and:

$$
\det A = 0
$$


Since each row sums to 1, there exists a **nonzero vector** $x$ such that:
$$
(A - I)x = 0
$$

so $A - I$ has a **nontrivial null space**, meaning:
$$
\det(A - I) = 0
$$


This does not mean $\det A = 1$. Since **determinants do not distribute over subtraction**, we cannot say:
$$
\det(A - I) = \det(A) - \det(I)
$$

Instead, $A$ could still have a **zero determinant** if it has a zero pivot. This means $A - I$ being singular tells us **nothing about $ \det A$**.
