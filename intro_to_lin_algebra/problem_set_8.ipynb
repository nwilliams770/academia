### (a) Show that \( \|u(t)\|^2 \) stays constant

Compute the derivative of the squared magnitude of \( u(t) \):

$$
\|u(t)\|^2 = u_1^2 + u_2^2 + u_3^2
$$

Differentiating using the chain rule:

$$
\frac{d}{dt} \|u(t)\|^2 = 2u_1 u_1' + 2u_2 u_2' + 2u_3 u_3'
$$

From the system:

- \( u_1' = cu_2 - bu_3 \)
- \( u_2' = au_3 - cu_1 \)
- \( u_3' = bu_1 - au_2 \)

Substituting:

\[
\begin{aligned}
2u_1 u_1' + 2u_2 u_2' + 2u_3 u_3' 
&= 2u_1(cu_2 - bu_3) + 2u_2(au_3 - cu_1) + 2u_3(bu_1 - au_2) \\
&= 2cu_1 u_2 - 2bu_1 u_3 + 2au_2 u_3 - 2cu_1 u_2 + 2bu_1 u_3 - 2au_2 u_3 \\
&= 0
\end{aligned}
\]

So:

$$
\frac{d}{dt} \|u(t)\|^2 = 0 \Rightarrow \|u(t)\| = \|u(0)\| \text{ for all } t
$$

This means the solution vector preserves its length — it moves on the surface of a sphere in \( \mathbb{R}^3 \). There is **no net scaling** ever — not even momentarily.

We are given:

$$
A = \begin{bmatrix} 1 & 1 \\ 0 & 3 \end{bmatrix}
$$

---

We recognize that \( A \) is upper triangular, so the eigenvalues are just the diagonal entries:

$$
\lambda_1 = 1, \quad \lambda_2 = 3
$$

---

To find the eigenvectors, solve $(A - \lambda I)x = 0$ for each eigenvalue.

**For $\lambda = 1$:**

$$
A - I = \begin{bmatrix} 0 & 1 \\ 0 & 2 \end{bmatrix}
$$

Solve:

$$
\begin{bmatrix} 0 & 1 \\ 0 & 2 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix} \Rightarrow y = 0
$$

Pick \( x = 1 \). So eigenvector is:

$$
x_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

---

**For $\lambda = 3$:**

$$
A - 3I = \begin{bmatrix} -2 & 1 \\ 0 & 0 \end{bmatrix}
$$

Solve:

$$
-2x + y = 0 \Rightarrow y = 2x \Rightarrow x = 1 \Rightarrow y = 2
$$

So eigenvector is:

$$
x_2 = \begin{bmatrix} 1 \\ 2 \end{bmatrix}
$$

---

We now have the eigenbasis matrix $X$ and the diagonal matrix $ \Lambda $:

$$
X = \begin{bmatrix} 1 & 1 \\ 0 & 2 \end{bmatrix}, \quad
\Lambda = \begin{bmatrix} 1 & 0 \\ 0 & 3 \end{bmatrix}
$$

To compute the matrix exponential $ e^{At} $, we use:

$$
e^{At} = X e^{\Lambda t} X^{-1}
$$

Because $ \Lambda $ is diagonal, $ e^{\Lambda t} $ is just exponentiating each diagonal entry:

$$
e^{\Lambda t} = \begin{bmatrix} e^t & 0 \\ 0 & e^{3t} \end{bmatrix}
$$

To compute \( X^{-1} \), use the 2×2 inverse formula:

$$
X^{-1} = \begin{bmatrix} 1 & -\tfrac{1}{2} \\ 0 & \tfrac{1}{2} \end{bmatrix}
$$

Now multiply:

$$
e^{At} = 
\begin{bmatrix} 1 & 1 \\ 0 & 2 \end{bmatrix}
\begin{bmatrix} e^t & 0 \\ 0 & e^{3t} \end{bmatrix}
\begin{bmatrix} 1 & -\tfrac{1}{2} \\ 0 & \tfrac{1}{2} \end{bmatrix}
=
\begin{bmatrix}
e^t & \tfrac{1}{2}(e^{3t} - e^t) \\
0 & e^{3t}
\end{bmatrix}
$$

28.

Given:
$$
A = \begin{bmatrix} 1 & 1 \\ -1 & 0 \end{bmatrix}
$$

To find the eigenvalues, solve:
$$
\det(A - \lambda I) = 0
$$

Compute the characteristic polynomial:
$$
\det \begin{bmatrix} 1 - \lambda & 1 \\ -1 & -\lambda \end{bmatrix}
= (1 - \lambda)(-\lambda) - (-1)(1)
= \lambda^2 - \lambda + 1 = 0
$$

Solve:
$$
\lambda = \frac{1 \pm \sqrt{-3}}{2} = \frac{1 \pm i\sqrt{3}}{2}
$$

So the eigenvalues are:
$$
\lambda_1 = \frac{1 + i\sqrt{3}}{2}, \quad
\lambda_2 = \frac{1 - i\sqrt{3}}{2}
$$

For \( \lambda_1 \), solve:
$$
(A - \lambda_1 I) \vec{x} = 0
$$

We use the first row:
$$
(1 - \lambda_1)x + y = 0 \Rightarrow y = (\lambda_1 - 1)x
$$

Let \( x = 1 \), then:
$$
\vec{x}_1 = \begin{bmatrix} 1 \\ \lambda_1 - 1 \end{bmatrix}
= \begin{bmatrix} 1 \\ \frac{-1 + i\sqrt{3}}{2} \end{bmatrix}
$$

Similarly, for \( \lambda_2 \), we get:
$$
\vec{x}_2 = \begin{bmatrix} 1 \\ \lambda_2 - 1 \end{bmatrix}
= \begin{bmatrix} 1 \\ \frac{-1 - i\sqrt{3}}{2} \end{bmatrix}
$$

Let:
$$
X = \begin{bmatrix}
1 & 1 \\
\frac{-1 + i\sqrt{3}}{2} & \frac{-1 - i\sqrt{3}}{2}
\end{bmatrix}, \quad
\Lambda = \begin{bmatrix}
\lambda_1 & 0 \\
0 & \lambda_2
\end{bmatrix}
$$

Then:
$$
A = X \Lambda X^{-1}, \quad A^6 = X \Lambda^6 X^{-1}
$$

Since both eigenvalues lie on the unit circle (magnitude 1), raising them to the 6th power rotates them full circle:
$$
\lambda_1^6 = \lambda_2^6 = 1
$$

So:
$$
\Lambda^6 = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix} = I
\Rightarrow A^6 = X I X^{-1} = I
$$

Final result:
$$
A^6 = I
$$

This confirms that the system cycles every 6 steps — it's a discrete rotation with period 6.

29.

We are given two matrices corresponding to leapfrog steps $\Delta t = \sqrt{2}$ and $\Delta t = 2$. We’re asked to:

- Find their eigenvalues  
- Compute $A^4$  
- Comment on stability (i.e. whether $|\lambda| \leq 1$)

### Matrix A₁: $\Delta t = \sqrt{2}$

$$
A_1 = \begin{bmatrix}
1 & \sqrt{2} \\
-\sqrt{2} & -1
\end{bmatrix}
$$


$$
\text{det}(A - \lambda I) =
\begin{vmatrix}
1 - \lambda & \sqrt{2} \\
-\sqrt{2} & -1 - \lambda
\end{vmatrix}
= (1 - \lambda)(-1 - \lambda) + 2
= -1 - \lambda^2 + 2 = -\lambda^2 + 1
$$

$$
\Rightarrow \lambda^2 = 1 \Rightarrow \lambda = \pm i
$$

So:
- Eigenvalues: $\lambda = i, -i$
- $|\lambda| = 1$ → stable (pure rotation)

#### Compute $A_1^2$

$$
A_1^2 = A_1 \cdot A_1 =
\begin{bmatrix}
1 & \sqrt{2} \\
-\sqrt{2} & -1
\end{bmatrix}
\begin{bmatrix}
1 & \sqrt{2} \\
-\sqrt{2} & -1
\end{bmatrix}
=
\begin{bmatrix}
1 - 2 & \sqrt{2} - \sqrt{2} \\
- \sqrt{2} + \sqrt{2} & -2 + 1
\end{bmatrix}
=
\begin{bmatrix}
-1 & 0 \\
0 & -1
\end{bmatrix}
= -I
$$

$$
A_1^4 = (A_1^2)^2 = (-I)^2 = I
$$

So $A_1^4 = I$, perfectly stable.

---

### Matrix A₂: $\Delta t = 2$

$$
A_2 = \begin{bmatrix}
1 & 2 \\
-2 & -3
\end{bmatrix}
$$

#### Characteristic Polynomial

$$
\text{det}(A - \lambda I) =
\begin{vmatrix}
1 - \lambda & 2 \\
-2 & -3 - \lambda
\end{vmatrix}
= (1 - \lambda)(-3 - \lambda) + 4
= -3 - \lambda + 3\lambda + \lambda^2 + 4 = \lambda^2 + 2\lambda + 1 = (\lambda + 1)^2
$$

So:
- Repeated eigenvalue: $\lambda = -1$
- $|\lambda| = 1$, but repeated roots with defective matrix → risk of instability

#### Compute $A_2^2$

$$
A_2^2 =
\begin{bmatrix}
1 & 2 \\
-2 & -3
\end{bmatrix}
\begin{bmatrix}
1 & 2 \\
-2 & -3
\end{bmatrix}
=
\begin{bmatrix}
1 - 4 & 2 - 6 \\
-2 + 6 & -4 + 9
\end{bmatrix}
=
\begin{bmatrix}
-3 & -4 \\
4 & 5
\end{bmatrix}
$$

#### Compute $A_2^4 = (A_2^2)^2$

$$
A_2^4 =
\begin{bmatrix}
-3 & -4 \\
4 & 5
\end{bmatrix}
\begin{bmatrix}
-3 & -4 \\
4 & 5
\end{bmatrix}
=
\begin{bmatrix}
9 - 16 & 12 - 20 \\
-12 + 20 & -16 + 25
\end{bmatrix}
=
\begin{bmatrix}
-7 & -8 \\
8 & 9
\end{bmatrix}
$$

So:

$$
A_2^4 = \begin{bmatrix}
-7 & -8 \\
8 & 9
\end{bmatrix} \neq I
$$

This shows a growing pattern → unstable dynamics even though $|\lambda| = 1$

---

### Summary

- **$A_1$ ($\Delta t = \sqrt{2}$)**:
  - $\lambda = \pm i$, unit circle, rotation  
  - $A_1^4 = I$ → stable

- **$A_2$ ($\Delta t = 2$)**:
  - $\lambda = -1$, defective eigenbasis  
  - $A_2^4$ grows in magnitude → unstable

This illustrates the importance of spectral radius and matrix diagonalizability in system stability.


30.

We are given a system that uses the trapezoidal (half-forward, half-back) method:

$$
\begin{bmatrix}
1 & -\frac{\Delta t}{2} \\
\frac{\Delta t}{2} & 1
\end{bmatrix}
\begin{bmatrix}
Y_{n+1} \\
Z_{n+1}
\end{bmatrix}
=
\begin{bmatrix}
1 & \frac{\Delta t}{2} \\
-\frac{\Delta t}{2} & 1
\end{bmatrix}
\begin{bmatrix}
Y_n \\
Z_n
\end{bmatrix}
$$

We want to write this as a first-order update:  
$$U_{n+1} = A U_n$$  
where  
$$A = M^{-1}N$$  
and  
$$
M = \begin{bmatrix}
1 & -\frac{\Delta t}{2} \\
\frac{\Delta t}{2} & 1
\end{bmatrix}, \quad
N = \begin{bmatrix}
1 & \frac{\Delta t}{2} \\
-\frac{\Delta t}{2} & 1
\end{bmatrix}
$$

### Invert $M$

Compute determinant of $M$:
$$
\det(M) = 1^2 + \left( \frac{\Delta t}{2} \right)^2 = 1 + \frac{\Delta t^2}{4}
$$

Then:
$$
M^{-1} = \frac{1}{1 + \frac{\Delta t^2}{4}} \begin{bmatrix}
1 & \frac{\Delta t}{2} \\
-\frac{\Delta t}{2} & 1
\end{bmatrix}
$$

### Multiply to find $A$

Now compute:
$$
A = M^{-1}N = \frac{1}{1 + \frac{\Delta t^2}{4}} \begin{bmatrix}
1 & \frac{\Delta t}{2} \\
-\frac{\Delta t}{2} & 1
\end{bmatrix}
\begin{bmatrix}
1 & \frac{\Delta t}{2} \\
-\frac{\Delta t}{2} & 1
\end{bmatrix}
$$

This results in a matrix $A$ such that:
$$
A^\top A = I
$$

Therefore, **$A$ is an orthogonal matrix**.

- $A$ is orthogonal: it preserves lengths and angles.
- So $||U_n|| = 1$ for all $n$: the update stays exactly on the unit circle.
- This method has **perfect numerical stability** for $y'' = -y$.

If $$A^3 = 0$$, then all eigenvalues of $$A$$ must be zero.

Any matrix $$A$$ with eigenvector $$x$$ and eigenvalue $$\lambda$$:

$$
Ax = \lambda x \Rightarrow A^3 x = \lambda^3 x
$$

If $$A^3 = 0$$, then $$A^3 x = 0$$ for all $$x$$. This implies:

$$
\lambda^3 x = 0 \Rightarrow \lambda^3 = 0 \Rightarrow \lambda = 0
$$

So all eigenvalues of $$A$$ must be zero.

---

Let

$$
A = \begin{bmatrix}
0 & 1 \\
0 & 0
\end{bmatrix}
$$

Then:

$$
A^2 = \begin{bmatrix}
0 & 0 \\
0 & 0
\end{bmatrix}, \quad A^3 = 0
$$

But $$A \neq 0$$. This shows a nonzero matrix can still satisfy $$A^3 = 0$$.

---

If $$A$$ is symmetric, then it can be diagonalized as:

$$
A = Q \Lambda Q^T
$$

Then:

$$
A^3 = Q \Lambda^3 Q^T
$$

If $$A^3 = 0$$, then:

$$
Q \Lambda^3 Q^T = 0 \Rightarrow \Lambda^3 = 0 \Rightarrow \Lambda = 0
$$

Hence:

$$
A = Q \cdot 0 \cdot Q^T = 0
$$

So the only **symmetric** matrix that satisfies $$A^3 = 0$$ is the **zero matrix**.

---

- All eigenvalues must be $$0$$.
- If $$A$$ is not symmetric, it might still be nonzero (e.g., nilpotent upper triangular matrix).
- If $$A$$ is symmetric, then $$A$$ must be the zero matrix.

By the spectral theorem, a real symmetric matrix can be diagonalized as

A = Q \Lambda Q^T
$$  
where $Q$ is orthogonal (i.e., columns are orthonormal eigenvectors). The converse also holds.

The inverse of a symmetric matrix is symmetric.

$$
(A^{-1})^T = (A^T)^{-1} = A^{-1}
$$  
So $A^{-1}$ is symmetric too.

**(d)** False.

The eigenvector matrix $Q$ is **orthogonal**, but not necessarily **symmetric**.
