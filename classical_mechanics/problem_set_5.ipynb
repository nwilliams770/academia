1a.

Because the collision is near-instantaneous, the external impulse from static friction is negligble and momentum is conserved across the collision:
$$
m1u = (m1 + m2)v
$$

Where $v$ is the speed of the billet-block after collision:
$$
v = \frac{m1u}{m1 + m2}
$$

As they move together, kinetic friction acts opposing the motion, we can calculate this force and then derive the deceleration from it
$$
F_{kf} = \mu_k(m1 + m2)g = (m1 + m2)a
$$

$$
a = \mu_kg
$$

Deceleration doesn't depend on the masses but instead on the coefficient of friction based on the surface!

The block's velocity at any time $t'$ after the collision is:
$$
v(t') = v - \mu_kgt'
$$

The min condition is that the block just barely reaches the edge before coming to rest. At rest we know $v = 0$ so we can use this to solve for $t'$:
$$
v - \mu_kgt' = 0 \implies t' = \frac{v}{\mu_kg}
$$

The distance traveled at that moment must be exactly $d$. We can integrate the velocity function to get the position function:
$$
vt' - \frac{1}{2} \mu_kgt'^2 = d \implies d = \frac{v^2}{2\mu_kg} \implies v = \sqrt{2 \mu_kgd}
$$

From here we can substitute and solve for our $u_min$:
$$
\frac{m1u_{min}}{m1 + m2} = \sqrt{2 \mu_kgd} \implies u_{min} = \frac{m1 + m2}{m1} \sqrt{2 \mu_kgd}
$$

1b.

After collision our block moves at $v = \frac{m1u}{m1 + m2}$ and decelerates due to kinetic friction over distance $d$. Using the velocity and position equations from part a, we can get the velocity when the block is at the edge:
$$
v_{edge} = v^2 - 2 \mu_k gd = (\frac{m1u}{m1 + m2})^2 - 2 \mu_k gd
$$

This is ONLY valid when $u >= u_min$ otherwise the block never reaches the edge.

The block leaves the edge with zero vertical velocity and falls height $h$ under gravity alone:
$$
h = \frac{1}{2}gt^2 \implies t = \frac{2h}{g}
$$

No horizontal forces act during the fall so finding $x_{final}$ is just $v_{edge}t$:
$$
x_{final} = v_{edge}t = \frac{2h}{g} [(\frac{m1u}{m1 + m2})^2 - 2 \mu_k gd)]
$$

2.

The acrobat launches at $v0$ and decelerates under gravity over height $h0$. Use standard kinematic equation $v_{final}^2 = v_{initial}^2 - 2ad$ to find $v_{h_0}$:
$$
v_{h_0} = v_0^2 - 2gh_0
$$

The grab is instantaneous so gravity delivers near zero impulse. Momentum is conserved, and we can use conservation of momentum to find the post-grab velocity:
$$
m_{acrobat} v_{h_0} = (m_{acrobat} + m_{clown})v_{new} \implies v_{new} = \frac{m_{acrobat}}{m_{acrobat} + m_{clown}} v_0^2 - 2gh_0
$$

The new system also decelerates under gravity from height $h0$. At maxiumum height the velocity is 0, we can use the kinematic equation again:
$$
0 = v_{new}^2 - 2g(h_{final} - h_0) \implies h_{final} = h0 + \frac{v_{new}^2}{2g}
$$

Substitute $v_{new}$:
$$
h_{final} = h0 + (\frac{m_{acrobat}}{m_{acrobat} + m_{clown}})^2 \frac{v_0^2 - 2gh_0}{2g}
$$

3a.

We can use the standard kinematic equation to find the speed of the person when they hit the ground, noting that they start from rest ($v_initial = 0$):
$$
v_{impact} = 2gh_0 = 2 \times 9.8 \times 2.0 \approx 6.3 \, m/s
$$

The decelerate constantly at a rate $a$ from $v_{impact}$ to zero over distance $d = 1.0 cm$. We have a velocity and position equation, starting from moment of impact.
Velocity is 0 by the end of collision:
$$
v_{impact} - a \Delta t_{collision} = 0
$$

Center of mass drops distance $d$
$$
v_{impact} \Delta t_{collision} - \frac{a \Delta t_{collision}^2}{2} = d
$$

We can solve for $a$ by using the standard kinematic equation again, knowing that $v_{final} = 0$:

$$
0 = v_{impact}^2 - 2ad \implies a = \frac{v_{impact}^2}{2d} = \frac{6.3^2}{2 \times 0.01} \approx 1984.5 \, m/s
$$

Now sub in from the velocity equation:
$$
\Delta t_{collision} = \frac{v_{impact}}{a} = \frac{6.3}{1984.5} \approx 3.2 \times 10^{-3} \, s
$$


3b.

$$
N_{average} = ma = 60 \times 1984.5 \approx 1.2 \times 10^5 \, N

\frac{1.2 \times 10^5}{5.9 \times 10^2} \approx 204

3d.

We find the max force the tibia can withstand before breaking by multiplying the threshold by cross sectional area:
$$
F_{break} = (1.6 \times 10^8 \, \frac{N}{m^2} ) \times (0.01 \, \frac{m}{cm})^2 \times (3.2 \, cm^2) \approx 5.1 \times 10^4 \, N

This is less than the ground force, so the person breaks their ankles.

4a. 

The linear mass density is constant as the rod is uniform therefore $\lambda = \frac{M}{L}$. Substituting into the center of mass integral:
$$
x_{center \, of \, mass} = \frac{1}{M} \int_{0}^{L} x \frac{M}{L} dx = \frac{1}{L} \int_{0}^{L} xdx = \frac{1}{L} \frac{L^2}{2} = \frac{L}{2}
$$

$r_{cm} = \frac{L}{2} \hat{i}$

4b.

First we solve for $\lambda_0$. We know that the total mass of the rod is just $M$ and the total is just the integral of density over the whole rod:
$$
M = \int_{0}^{L} \lambda dx = \int_{0}^{L} \frac{\lambda_0}{L^3} x^3 dx \implies \lambda_0 = \frac{4M}{L}
$$

Now we can solve for the center of mass of the rod:
$$
x_{center \, of \, mass} = \frac{1}{M} \int_{0}^{L} x \frac{lambda_0}{L^3} x^3 dx
$$

Substitute for $\lambda_0$, move the constants outside the integral, and simplify:
$$
x_{center \, of \, mass} = \frac{4}{L^4} \int_{0}^{L} x^4 dx = \frac{4}{L^4} \frac{L^5}{5} = \frac{4L}{5}
$$

$r_{cm} = \frac{4L}{5} \hat{i}$

We can define $x_1$ and $x_2$ as the positions of $m_1$ and $m_2$, respectively. At $t_0$, $m_1$ is at our origin:
$$
x_1(t_0) = 0 \, , \, x_2(t_0) = d  \, , \, v_1(t_0) = v_0  \, , \, v_2(t_0) = 0
$$

Center of mass position and velocity are weighted averages of the individual positions and velocities:
$$
x_{center \, of \, mass} = \frac{m_1x_1 + m_2x_2}{m_1 + m_2} \, , \, v_{center \, of \, mass} = \frac{m_1v_1 + m_2v_2}{m_1 + m_2}
$$

At $t_0$:
$$
x_{cm} = \frac{m_2d}{m_1 + m_2} \, , \, v_{cm} = \frac{m_1v_0}{m_1 + m_2} 
$$

The forces on each particle are equal and opposite therefore the acceleration of their center of mass is 0. This means that $v_{cm}$ is constant. At $t1$ the particles collide, occupying the same position. THat position must equal where the center of mass is at $t1$:

$$
x_1(t_1) = x_{cm}(t_1) = x_cm(t0) + v_{cm}t_1 = \frac{m_2d}{m_1 + m_2} + \frac{m_1v_0}{m_1 + m_2}t_1
$$
